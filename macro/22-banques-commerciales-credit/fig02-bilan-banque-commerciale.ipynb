```python
LANG = "fr"   # "fr" ou "en" — langue des libellés / label language

# Récupère puis active le style partagé NMLab (thème sombre + police Inter).
# Fetch and activate the shared NMLab style (dark theme + Inter font).
import urllib.request

urllib.request.urlretrieve("https://raw.githubusercontent.com/nmlab-finance/nmlab-figures/main/nmlab_style.py", "nmlab_style.py")
import nmlab_style as nm

nm.setup()


def balance_sheet(lang: str) -> tuple[list[dict], list[dict]]:
    """Blocs stylisés de l'actif et du passif (nom, part en %, style), localisés.
    Stylized asset and liability blocks (name, share in %, style), localized."""
    if lang == "fr":
        assets = [
            dict(name="Réserves en banque centrale", pct=10, color="amber", star=False),
            dict(name="Titres (obligations, etc.)", pct=20, color="blue2", star=False),
            dict(name="Crédits à la clientèle", pct=60, color="blue", star=True),
            dict(name="Autres actifs", pct=10, color="muted", star=False),
        ]
        liabs = [
            dict(name="Dépôts de la clientèle", pct=55, color="teal", star=True),
            dict(name="Financements de marché", pct=25, color="amber", star=False),
            dict(name="Autres passifs", pct=12, color="muted", star=False),
            dict(name="Fonds propres", pct=8, color="rose", star=True),
        ]
        return assets, liabs
    assets = [
        dict(name="Central-bank reserves", pct=10, color="amber", star=False),
        dict(name="Securities (bonds, etc.)", pct=20, color="blue2", star=False),
        dict(name="Loans to customers", pct=60, color="blue", star=True),
        dict(name="Other assets", pct=10, color="muted", star=False),
    ]
    liabs = [
        dict(name="Customer deposits", pct=55, color="teal", star=True),
        dict(name="Market funding", pct=25, color="amber", star=False),
        dict(name="Other liabilities", pct=12, color="muted", star=False),
        dict(name="Equity (capital)", pct=8, color="rose", star=True),
    ]
    return assets, liabs


from matplotlib.figure import Figure

LABELS = {
    "fr": dict(
        title="Le bilan d'une banque commerciale, en coupe",
        sub="Structure stylisée — à l'actif, ce que la banque détient ; au passif, ce qu'elle doit, dont ses fonds propres.",
        col_a="Actif", col_l="Passif",
        note="Proportions stylisées d'une grande banque universelle de la zone euro — elles varient selon les modèles.\n"
             "Les fonds propres absorbent les pertes ; les réserves règlent les paiements entre banques."),
    "en": dict(
        title="A commercial bank's balance sheet, in cross-section",
        sub="Stylized structure — assets: what the bank holds; liabilities: what it owes, including its capital.",
        col_a="Assets", col_l="Liabilities",
        note="Stylized proportions for a large euro-area universal bank — they vary across business models.\n"
             "Equity absorbs losses; reserves settle payments between banks."),
}

def build_figure(content: tuple[list[dict], list[dict]], lang: str) -> Figure:
    """Deux colonnes empilées (actif / passif), blocs proportionnels aux parts stylisées."""
    text = LABELS[lang]
    assets, liabs = content
    fig = nm.figure(height_px=1080)
    ax = nm.blank_axes(fig)

    cols = [(90, 800, text["col_a"], assets), (947, 1657, text["col_l"], liabs)]
    top, bottom, gap = 760, 170, 8
    span = top - bottom - gap * 3
    for x0, x1, col_title, blocks in cols:
        cx = (x0 + x1) / 2
        ax.text(cx, 806, col_title, ha="center", va="center", fontsize=27,
                fontweight="bold", color=nm.COLORS["text"])
        y = top
        for b in blocks:
            h = span * b["pct"] / 100
            color = nm.COLORS[b["color"]]
            if b["star"]:
                nm.card(ax, x0, y - h, x1 - x0, h, edge=color, fill=color, radius=14)
                name_c, pct_c = nm.COLORS["bg"], nm.COLORS["bg"]
            else:
                nm.card(ax, x0, y - h, x1 - x0, h, edge=color, radius=14, lw=2.2)
                name_c, pct_c = nm.COLORS["text"], nm.COLORS["muted"]
            fs = 24 if h >= 56 else 21
            ax.text(x0 + 28, y - h / 2, b["name"], ha="left", va="center",
                    fontsize=fs, fontweight="bold" if b["star"] else "normal", color=name_c)
            ax.text(x1 - 28, y - h / 2, f"≈ {b['pct']} %", ha="right", va="center",
                    fontsize=fs, fontweight="bold", color=pct_c)
            y -= h + gap

    nm.header(fig, text["title"], text["sub"])
    nm.footer(fig, text["note"])
    return fig

build_figure(balance_sheet(LANG), LANG)
```
