```python
LANG = "fr"   # "fr" ou "en" — langue des libellés / label language

# Récupère puis active le style partagé NMLab (thème sombre + police Inter).
# Fetch and activate the shared NMLab style (dark theme + Inter font).
import urllib.request

urllib.request.urlretrieve("https://raw.githubusercontent.com/nmlab-finance/nmlab-figures/main/nmlab_style.py", "nmlab_style.py")
import nmlab_style as nm

nm.setup()


def dials(lang: str) -> list[dict]:
    """Les quatre cadrans du crédit (titre, corps, couleur), localisés.
    The four credit dials (title, body, color), localized."""
    if lang == "fr":
        return [
            dict(title="Encours et croissance du crédit", color="blue",
                 body="Prêts aux ménages et aux entreprises,\npubliés chaque mois (BCE, Webstat\nBanque de France) : le pouls de la\ncréation monétaire privée."),
            dict(title="Conditions d'octroi", color="amber",
                 body="Bank Lending Survey trimestrielle de\nla BCE, SLOOS de la Fed : le durcissement\ndes critères précède souvent le cycle\ndu crédit — et l'activité."),
            dict(title="Taux des nouveaux crédits", color="teal",
                 body="Habitat en France : ≈ 1,1 % fin 2021,\nplus de 4 % début 2024 — et une\nproduction de prêts divisée par près\nde trois dans l'intervalle."),
            dict(title="Impulsion de crédit", color="rose",
                 body="La variation des flux nouveaux de\ncrédit, rapportée au PIB : souvent en\navance sur la demande intérieure et\nles actifs cycliques."),
        ]
    return [
        dict(title="Credit stock and growth", color="blue",
             body="Loans to households and firms,\npublished monthly (ECB, Banque de\nFrance Webstat): the pulse of private\nmoney creation."),
        dict(title="Lending standards", color="amber",
             body="The ECB's quarterly Bank Lending\nSurvey, the Fed's SLOOS: tightening\nstandards often lead the credit\ncycle — and activity."),
        dict(title="Rates on new loans", color="teal",
             body="French housing loans: ≈ 1.1% in late\n2021, above 4% in early 2024 — with\nnew lending divided by nearly three\nin between."),
        dict(title="Credit impulse", color="rose",
             body="The change in new credit flows,\nscaled by GDP: often ahead of\ndomestic demand and cyclical\nassets."),
    ]


from matplotlib.figure import Figure

LABELS = {
    "fr": dict(
        title="Le tableau de bord du crédit",
        sub="Quatre cadrans pour lire l'offre et la demande de crédit — et anticiper leurs effets sur l'activité et les marchés.",
        note="Le crédit est un indicateur avancé imparfait, mais son retournement a accompagné la plupart des grands\n"
             "cycles immobiliers et boursiers. Le chapitre 17 assemblera le tableau de bord complet."),
    "en": dict(
        title="The credit dashboard",
        sub="Four dials to read credit supply and demand — and anticipate their effects on activity and markets.",
        note="Credit is an imperfect leading indicator, but its turning points have accompanied most major housing\n"
             "and equity cycles. Chapter 17 will assemble the full dashboard."),
}

def build_figure(cards: list[dict], lang: str) -> Figure:
    """Grille 2 × 2 de cartes, une par cadran."""
    text = LABELS[lang]
    fig = nm.figure(height_px=1080)
    ax = nm.blank_axes(fig)

    xs, ys = [65, 902], [480, 120]
    for i, c in enumerate(cards):
        x, y = xs[i % 2], ys[i // 2]
        color = nm.COLORS[c["color"]]
        nm.card(ax, x, y, 780, 340, edge=color, lw=2.4, radius=22)
        ax.text(x + 42, y + 284, c["title"], ha="left", va="center", fontsize=27,
                fontweight="bold", color=color)
        ax.text(x + 42, y + 206, c["body"], ha="left", va="top", fontsize=22.5,
                color=nm.COLORS["muted"], linespacing=1.42)

    nm.header(fig, text["title"], text["sub"])
    nm.footer(fig, text["note"])
    return fig

build_figure(dials(LANG), LANG)
```
