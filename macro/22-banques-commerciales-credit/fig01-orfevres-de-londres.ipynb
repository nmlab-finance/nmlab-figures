```python
LANG = "fr"   # "fr" ou "en" — langue des libellés / label language

# Récupère puis active le style partagé NMLab (thème sombre + police Inter).
# Fetch and activate the shared NMLab style (dark theme + Inter font).
import urllib.request

urllib.request.urlretrieve("https://raw.githubusercontent.com/nmlab-finance/nmlab-figures/main/nmlab_style.py", "nmlab_style.py")
import nmlab_style as nm

nm.setup()


def goldsmith_cards(lang: str) -> list[tuple[str, str]]:
    """Les trois étapes de la découverte des orfèvres : (titre, corps), localisées.
    The three steps of the goldsmiths' discovery: (title, body), localized."""
    if lang == "fr":
        return [
            ("L'or en dépôt",
             "Les marchands confient leur\nor aux orfèvres ; le reçu\nprouve le dépôt et promet\nla restitution à vue."),
            ("Le reçu circule",
             "Payable au porteur, le reçu\npaie à son tour : l'or dort\nau coffre, le papier change\nde mains."),
            ("Le pari de l'orfèvre",
             "Une fraction seulement des\nreçus revient au coffre :\nl'orfèvre prête en émettant\nplus de reçus que d'or."),
        ]
    return [
        ("Gold on deposit",
         "Merchants entrust their gold\nto goldsmiths; the receipt\nproves the deposit, payable\non demand."),
        ("The receipt circulates",
         "Payable to the bearer, the\nreceipt itself pays: gold\nsleeps in the vault, paper\nchanges hands."),
        ("The goldsmith's bet",
         "Only a fraction of receipts\ncomes back to the vault: the\ngoldsmith lends by issuing\nmore receipts than gold."),
    ]


from matplotlib.figure import Figure

LABELS = {
    "fr": dict(
        title="Du reçu d'orfèvre au billet : le crédit fait la monnaie",
        sub="Londres, 1640-1685 : la saisie royale pousse l'or chez les orfèvres ; leurs reçus deviennent moyens de paiement.",
        left_date="1640", left_sub="Saisie royale à la Tour",
        right_date="v. 1685", right_sub="Le reçu devient billet",
        pill="≈ un demi-siècle",
        note="La saisie de 1640 brise la confiance dans la garde royale ; l'orfèvre découvre qu'une encaisse fractionnaire\n"
             "suffit — prêter, c'est émettre des reçus qui circulent comme monnaie."),
    "en": dict(
        title="From goldsmith's receipt to banknote: credit makes money",
        sub="London, 1640-1685: the royal seizure pushes gold to the goldsmiths; their receipts become means of payment.",
        left_date="1640", left_sub="Royal seizure at the Tower",
        right_date="c. 1685", right_sub="The receipt becomes a note",
        pill="≈ half a century",
        note="The 1640 seizure breaks trust in royal safekeeping; the goldsmith discovers that a fractional till is\n"
             "enough — to lend is to issue receipts that circulate as money."),
}

def build_figure(cards: list[tuple[str, str]], lang: str) -> Figure:
    """Frise 1640 → v. 1685 (dot rose → dot bleu) puis trois cartes de la découverte."""
    text = LABELS[lang]
    fig = nm.figure(height_px=1040)
    ax = nm.blank_axes(fig)

    # Frise : ligne, deux jalons, pastille centrale.
    y = 695
    ax.plot([205, 1540], [y, y], color=nm.COLORS["edge"], linewidth=3, zorder=1)
    ax.scatter([178], [y], s=300, color=nm.COLORS["rose"], zorder=3, clip_on=False)
    ax.scatter([1568], [y], s=300, color=nm.COLORS["blue"], zorder=3, clip_on=False)
    nm.card(ax, 714, y - 39, 320, 78, edge=nm.COLORS["blue"], radius=39, lw=2.4)
    ax.text(874, y, text["pill"], ha="center", va="center", fontsize=27,
            fontweight="bold", color=nm.COLORS["blue"], zorder=4)
    ax.text(90, y + 62, text["left_date"], ha="left", va="center", fontsize=28,
            fontweight="bold", color=nm.COLORS["text"])
    ax.text(120, y - 62, text["left_sub"], ha="left", va="center", fontsize=23, color=nm.COLORS["muted"])
    ax.text(1660, y + 62, text["right_date"], ha="right", va="center", fontsize=28,
            fontweight="bold", color=nm.COLORS["text"])
    ax.text(1628, y - 62, text["right_sub"], ha="right", va="center", fontsize=23, color=nm.COLORS["muted"])

    # Trois cartes de la découverte.
    card_w, gap, x0 = 515, 35, 65
    top, bottom = 565, 185
    for i, (title, body) in enumerate(cards):
        x = x0 + i * (card_w + gap)
        nm.card(ax, x, bottom, card_w, top - bottom, edge=nm.COLORS["edge"], lw=2.2, radius=22)
        ax.text(x + 42, top - 56, title, ha="left", va="center", fontsize=30,
                fontweight="bold", color=nm.COLORS["text"])
        ax.text(x + 42, top - 140, body, ha="left", va="top", fontsize=23,
                color=nm.COLORS["muted"], linespacing=1.45)

    nm.header(fig, text["title"], text["sub"])
    nm.footer(fig, text["note"])
    return fig

build_figure(goldsmith_cards(LANG), LANG)
```
