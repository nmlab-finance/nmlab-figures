```python
LANG = "fr"   # "fr" ou "en" — langue des libellés / label language

# Récupère puis active le style partagé NMLab (thème sombre + police Inter).
# Fetch and activate the shared NMLab style (dark theme + Inter font).
import urllib.request

urllib.request.urlretrieve("https://raw.githubusercontent.com/nmlab-finance/nmlab-figures/main/nmlab_style.py", "nmlab_style.py")
import nmlab_style as nm

nm.setup()


def circuits(lang: str) -> list[dict]:
    """Les deux circuits de financement (banque / marché), localisés.
    The two financing circuits (bank / market), localized."""
    if lang == "fr":
        return [
            dict(title="Crédit bancaire — création", color="blue",
                 top="Banque", bottom="Entreprise",
                 arrow="prête 100 :\ncrée un dépôt\nde 100",
                 pill="M3 : + 100", pill_color="green"),
            dict(title="Marché — transfert", color="amber",
                 top="Épargnant / fonds", bottom="Entreprise",
                 arrow="achète l'obligation :\nun dépôt existant\nchange de mains",
                 pill="M3 : inchangé", pill_color="amber"),
        ]
    return [
        dict(title="Bank credit — creation", color="blue",
             top="Bank", bottom="Firm",
             arrow="lends 100:\ncreates a deposit\nof 100",
             pill="M3: + 100", pill_color="green"),
        dict(title="Market — transfer", color="amber",
             top="Saver / fund", bottom="Firm",
             arrow="buys the bond:\nan existing deposit\nchanges hands",
             pill="M3: unchanged", pill_color="amber"),
    ]


from matplotlib.figure import Figure

LABELS = {
    "fr": dict(
        title="Qui crée la monnaie, qui la déplace",
        sub="Le prêt bancaire ajoute un dépôt neuf à la masse monétaire ; l'achat d'une obligation fait circuler un dépôt existant.",
        note="Marchés obligataires, fonds de dette et financement participatif transfèrent une monnaie déjà créée ;\n"
             "seules les banques — et la banque centrale — en créent. D'où leur régime prudentiel à part."),
    "en": dict(
        title="Who creates money, who moves it",
        sub="A bank loan adds a brand-new deposit to the money stock; buying a bond makes an existing deposit circulate.",
        note="Bond markets, debt funds and crowdfunding transfer money already created; only banks — and the\n"
             "central bank — create it. Hence their prudential regime apart."),
}

def build_figure(panels: list[dict], lang: str) -> Figure:
    """Deux panneaux : acteur du haut → flèche annotée → entreprise, puis pastille M3."""
    text = LABELS[lang]
    fig = nm.figure(height_px=1080)
    ax = nm.blank_axes(fig)

    for i, p in enumerate(panels):
        x0 = 65 + i * 837
        x1 = x0 + 780
        cx = (x0 + x1) / 2
        color = nm.COLORS[p["color"]]
        nm.card(ax, x0, 150, 780, 670, edge=color, lw=2.4, radius=22)
        ax.text(cx, 772, p["title"], ha="center", va="center", fontsize=27,
                fontweight="bold", color=color)
        nm.card(ax, cx - 190, 620, 380, 92, edge=nm.COLORS["edge"], lw=2.2, radius=18)
        ax.text(cx, 666, p["top"], ha="center", va="center", fontsize=25, color=nm.COLORS["text"])
        nm.card(ax, cx - 190, 380, 380, 92, edge=nm.COLORS["edge"], lw=2.2, radius=18)
        ax.text(cx, 426, p["bottom"], ha="center", va="center", fontsize=25, color=nm.COLORS["text"])
        ax.annotate("", xy=(cx - 120, 486), xytext=(cx - 120, 606),
                    arrowprops=dict(arrowstyle="-|>", color=color, lw=3))
        ax.text(cx - 74, 546, p["arrow"], ha="left", va="center", fontsize=22,
                color=nm.COLORS["muted"], linespacing=1.45)
        pill_c = nm.COLORS[p["pill_color"]]
        nm.card(ax, cx - 160, 224, 320, 74, edge=pill_c, radius=37, lw=2.4)
        ax.text(cx, 261, p["pill"], ha="center", va="center", fontsize=26,
                fontweight="bold", color=pill_c)

    nm.header(fig, text["title"], text["sub"])
    nm.footer(fig, text["note"])
    return fig

build_figure(circuits(LANG), LANG)
```
