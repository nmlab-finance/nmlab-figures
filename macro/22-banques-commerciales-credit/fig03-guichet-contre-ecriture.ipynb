```python
LANG = "fr"   # "fr" ou "en" — langue des libellés / label language

# Récupère puis active le style partagé NMLab (thème sombre + police Inter).
# Fetch and activate the shared NMLab style (dark theme + Inter font).
import urllib.request

urllib.request.urlretrieve("https://raw.githubusercontent.com/nmlab-finance/nmlab-figures/main/nmlab_style.py", "nmlab_style.py")
import nmlab_style as nm

nm.setup()


def models(lang: str) -> dict:
    """Libellés des deux modèles (guichet faux / écriture vraie), localisés.
    Labels of the two models (false teller / true bookkeeping), localized."""
    if lang == "fr":
        return dict(
            wrong_boxes=["Épargnant", "Banque", "Emprunteur"],
            bank="Banque", household="Ménage",
            asset="Actif", liab="Passif",
            bank_a="+ Prêt 200 000 €", bank_l="+ Dépôt 200 000 €",
            hh_a="+ Dépôt 200 000 €", hh_l="+ Dette 200 000 €",
        )
    return dict(
        wrong_boxes=["Saver", "Bank", "Borrower"],
        bank="Bank", household="Household",
        asset="Assets", liab="Liabilities",
        bank_a="+ Loan €200,000", bank_l="+ Deposit €200,000",
        hh_a="+ Deposit €200,000", hh_l="+ Debt €200,000",
    )


from matplotlib.figure import Figure

LABELS = {
    "fr": dict(
        title="200 000 € : l'écriture qui crée la monnaie",
        sub="Deux bilans s'allongent simultanément — la banque acquiert une créance, le ménage reçoit un dépôt tout neuf.",
        wrong="Le modèle du guichet — faux",
        wrong_sub="Attendre l'épargne d'abord, prêter ensuite : ce n'est pas ainsi que naît le crédit.",
        right="Le modèle comptable — vrai : l'écriture crée le dépôt",
        right_sub="Aucun euro n'a quitté un autre compte : le dépôt est neuf.",
        note="Le prêt n'est pas un transfert d'épargne existante : au décaissement, l'actif de la banque (créance) et son\n"
             "passif (dépôt) augmentent du même montant ; l'emprunteur gagne un dépôt et une dette — pas de richesse."),
    "en": dict(
        title="€200,000: the entry that creates money",
        sub="Two balance sheets lengthen at once — the bank acquires a claim, the household receives a brand-new deposit.",
        wrong="The teller model — false",
        wrong_sub="Wait for savings first, lend second: that is not how credit is born.",
        right="The bookkeeping model — true: the entry creates the deposit",
        right_sub="Not one euro left another account: the deposit is new.",
        note="A loan is not a transfer of pre-existing savings: at drawdown, the bank's asset (claim) and its liability\n"
             "(deposit) rise by the same amount; the borrower gains a deposit and a debt — not wealth."),
}

def build_figure(m: dict, lang: str) -> Figure:
    """Panneau rose (modèle du guichet, barré) puis panneau vert (double T-compte)."""
    text = LABELS[lang]
    fig = nm.figure(height_px=1150)
    ax = nm.blank_axes(fig)

    # Panneau du haut : le modèle faux.
    nm.card(ax, 65, 620, 1617, 285, edge=nm.COLORS["rose"], lw=2.6, radius=22)
    ax.text(107, 857, text["wrong"], ha="left", va="center", fontsize=28,
            fontweight="bold", color=nm.COLORS["rose"])
    boxes_x = [(150, 430), (720, 1000), (1290, 1570)]
    for (bx0, bx1), name in zip(boxes_x, m["wrong_boxes"]):
        nm.card(ax, bx0, 700, bx1 - bx0, 96, edge=nm.COLORS["edge"], lw=2.2, radius=18)
        ax.text((bx0 + bx1) / 2, 748, name, ha="center", va="center", fontsize=25,
                color=nm.COLORS["text"])
    for xa, xb in [(430, 720), (1000, 1290)]:
        ax.annotate("", xy=(xb - 14, 748), xytext=(xa + 14, 748),
                    arrowprops=dict(arrowstyle="-|>", color=nm.COLORS["muted"], lw=2.6))
    ax.text(873, 655, text["wrong_sub"], ha="center", va="center", fontsize=22,
            color=nm.COLORS["muted"])

    # Panneau du bas : le modèle vrai (double T-compte).
    nm.card(ax, 65, 175, 1617, 435, edge=nm.COLORS["green"], lw=2.6, radius=22)
    ax.text(107, 562, text["right"], ha="left", va="center", fontsize=28,
            fontweight="bold", color=nm.COLORS["green"])
    tables = [(140, 800, m["bank"], m["bank_a"], m["bank_l"], nm.COLORS["blue"], nm.COLORS["teal"]),
              (947, 1607, m["household"], m["hh_a"], m["hh_l"], nm.COLORS["teal"], nm.COLORS["rose"])]
    for x0, x1, name, entry_a, entry_l, ca, cl in tables:
        cx = (x0 + x1) / 2
        nm.card(ax, x0, 290, x1 - x0, 230, edge=nm.COLORS["edge"], lw=2.2, radius=18)
        ax.text(cx, 484, name, ha="center", va="center", fontsize=26,
                fontweight="bold", color=nm.COLORS["text"])
        ax.plot([x0 + 26, x1 - 26], [456, 456], color=nm.COLORS["edge"], linewidth=1.6)
        ax.plot([cx, cx], [310, 456], color=nm.COLORS["edge"], linewidth=1.6)
        ax.text((x0 + cx) / 2, 420, m["asset"], ha="center", va="center", fontsize=21,
                color=nm.COLORS["muted"])
        ax.text((cx + x1) / 2, 420, m["liab"], ha="center", va="center", fontsize=21,
                color=nm.COLORS["muted"])
        ax.text((x0 + cx) / 2, 355, entry_a, ha="center", va="center", fontsize=24,
                fontweight="bold", color=ca)
        ax.text((cx + x1) / 2, 355, entry_l, ha="center", va="center", fontsize=24,
                fontweight="bold", color=cl)
    ax.text(873, 232, text["right_sub"], ha="center", va="center", fontsize=22,
            color=nm.COLORS["muted"])

    nm.header(fig, text["title"], text["sub"])
    nm.footer(fig, text["note"])
    return fig

build_figure(models(LANG), LANG)
```
