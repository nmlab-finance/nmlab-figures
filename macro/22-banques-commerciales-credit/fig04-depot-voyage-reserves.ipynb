```python
LANG = "fr"   # "fr" ou "en" — langue des libellés / label language

# Récupère puis active le style partagé NMLab (thème sombre + police Inter).
# Fetch and activate the shared NMLab style (dark theme + Inter font).
import urllib.request

urllib.request.urlretrieve("https://raw.githubusercontent.com/nmlab-finance/nmlab-figures/main/nmlab_style.py", "nmlab_style.py")
import nmlab_style as nm

nm.setup()


def settlement(lang: str) -> dict:
    """Libellés du circuit acheteur → vendeur et de la contrainte de financement.
    Labels of the buyer → seller circuit and the funding constraint, localized."""
    if lang == "fr":
        return dict(
            bank_a="Banque A — l'acheteur", bank_b="Banque B — le vendeur",
            a1="Dépôt du client : − 200 000 €", a2="Réserves à la BCE : − 200 000 €",
            b1="Dépôt du client : + 200 000 €", b2="Réserves à la BCE : + 200 000 €",
            a_sub="perd une ressource", b_sub="reçoit une ressource",
            pill="Règlement T2", arrow="200 000 €\nde réserves",
            box_title="La contrainte de refinancement",
            box_body="La banque A doit remplacer la ressource partie : attirer des dépôts, emprunter\n"
                     "à d'autres banques, émettre des titres — ou laisser fondre son encours de prêts.\n"
                     "En pratique, des millions de paiements se croisent chaque jour : seules les\n"
                     "positions nettes se règlent en monnaie centrale.",
        )
    return dict(
        bank_a="Bank A — the buyer", bank_b="Bank B — the seller",
        a1="Customer deposit: − €200,000", a2="Reserves at the ECB: − €200,000",
        b1="Customer deposit: + €200,000", b2="Reserves at the ECB: + €200,000",
        a_sub="loses a resource", b_sub="gains a resource",
        pill="T2 settlement", arrow="€200,000\nof reserves",
        box_title="The refinancing constraint",
        box_body="Bank A must replace the departed resource: attract deposits, borrow from\n"
                 "other banks, issue securities — or let its loan book shrink.\n"
                 "In practice, millions of payments cross every day: only net\n"
                 "positions settle in central-bank money.",
    )


from matplotlib.figure import Figure

LABELS = {
    "fr": dict(
        title="Le dépôt voyage, les réserves suivent",
        sub="L'emprunteur paie un vendeur client d'une autre banque : le dépôt migre, le règlement passe par la monnaie centrale.",
        note="Ce qu'une banque crée en prêtant peut la quitter au premier paiement : prêter engage à financer.\n"
             "À l'échelle du système, les fuites des unes sont les ressources des autres."),
    "en": dict(
        title="The deposit travels, reserves follow",
        sub="The borrower pays a seller who banks elsewhere: the deposit migrates, settlement goes through central-bank money.",
        note="What a bank creates by lending can leave it at the first payment: lending commits it to funding.\n"
             "At the system level, one bank's leakages are another's resources."),
}

def build_figure(s: dict, lang: str) -> Figure:
    """Deux cartes bancaires reliées par le règlement T2, puis la carte de financement."""
    text = LABELS[lang]
    fig = nm.figure(height_px=1150)
    ax = nm.blank_axes(fig)

    banks = [(65, 745, s["bank_a"], s["a1"], s["a2"], s["a_sub"], nm.COLORS["rose"]),
             (1002, 1682, s["bank_b"], s["b1"], s["b2"], s["b_sub"], nm.COLORS["green"])]
    for x0, x1, name, l1, l2, sub, accent in banks:
        cx = (x0 + x1) / 2
        nm.card(ax, x0, 555, x1 - x0, 350, edge=nm.COLORS["edge"], lw=2.4, radius=22)
        ax.text(cx, 855, name, ha="center", va="center", fontsize=27,
                fontweight="bold", color=nm.COLORS["text"])
        ax.text(x0 + 48, 775, l1, ha="left", va="center", fontsize=24, color=accent)
        ax.text(x0 + 48, 712, l2, ha="left", va="center", fontsize=24, color=nm.COLORS["amber"])
        ax.text(cx, 615, sub, ha="center", va="center", fontsize=22,
                color=nm.COLORS["muted"], style="italic")

    # Règlement central : pastille T2 + flèche de réserves.
    nm.card(ax, 756, 780, 235, 62, edge=nm.COLORS["blue"], radius=31, lw=2.4)
    ax.text(873, 811, s["pill"], ha="center", va="center", fontsize=23,
            fontweight="bold", color=nm.COLORS["blue"], zorder=4)
    ax.annotate("", xy=(988, 660), xytext=(759, 660),
                arrowprops=dict(arrowstyle="-|>", color=nm.COLORS["blue"], lw=3))
    ax.text(873, 712, s["arrow"], ha="center", va="center", fontsize=21,
            color=nm.COLORS["muted"], linespacing=1.4)

    # Carte du bas : la contrainte de refinancement.
    nm.card(ax, 65, 160, 1617, 305, edge=nm.COLORS["amber"], lw=2.4, radius=22)
    ax.text(107, 417, s["box_title"], ha="left", va="center", fontsize=26,
            fontweight="bold", color=nm.COLORS["amber"])
    ax.text(107, 358, s["box_body"], ha="left", va="top", fontsize=22.5,
            color=nm.COLORS["muted"], linespacing=1.42)

    nm.header(fig, text["title"], text["sub"])
    nm.footer(fig, text["note"])
    return fig

build_figure(settlement(LANG), LANG)
```
