```python
LANG = "fr"   # "fr" ou "en" — langue des libellés / label language

# Récupère puis active le style partagé NMLab (thème sombre + police Inter).
# Fetch and activate the shared NMLab style (dark theme + Inter font).
import urllib.request

urllib.request.urlretrieve("https://raw.githubusercontent.com/nmlab-finance/nmlab-figures/main/nmlab_style.py", "nmlab_style.py")
import nmlab_style as nm

nm.setup()


def flows(lang: str) -> tuple[list[dict], list[tuple[str, str]]]:
    """Barres de flux stylisés (base 100) et deux épisodes historiques, localisés.
    Stylized flow bars (base 100) and two historical episodes, localized."""
    if lang == "fr":
        bars = [
            dict(name="Crédits nouveaux", value=100, label="100", color="blue"),
            dict(name="Remboursements", value=85, label="− 85", color="rose"),
            dict(name="Création nette", value=15, label="+ 15", color="green"),
        ]
        cases = [
            ("2020-2021 — la vague des PGE",
             "≈ 143 Md€ garantis à près de\n700 000 entreprises : les dépôts\ndes PME bondissent d'autant."),
            ("2013-2014 — le reflux",
             "Crédit aux entreprises : ≈ − 3 %\nsur un an début 2014 ; la création\nnette devient négative en zone euro."),
        ]
        return bars, cases
    bars = [
        dict(name="New loans", value=100, label="100", color="blue"),
        dict(name="Repayments", value=85, label="− 85", color="rose"),
        dict(name="Net creation", value=15, label="+ 15", color="green"),
    ]
    cases = [
        ("2020-2021 — the French PGE wave",
         "≈ €143bn of state-guaranteed loans\nto nearly 700,000 firms: SME\ndeposits jump by as much."),
        ("2013-2014 — the ebb",
         "Loans to firms: ≈ − 3% year on year\nin early 2014; net creation turns\nnegative in the euro area."),
    ]
    return bars, cases


from matplotlib.figure import Figure

LABELS = {
    "fr": dict(
        title="La monnaie respire : brut, remboursé, net",
        sub="Flux d'une année, stylisés (base 100) : les crédits nouveaux créent des dépôts, les remboursements en détruisent.",
        note="Le remboursement du principal éteint le dépôt qui lui fait face ; les intérêts, eux, sont le revenu\n"
             "de la banque — ils changent de mains sans être détruits. La masse monétaire suit le solde net."),
    "en": dict(
        title="Money breathes: gross, repaid, net",
        sub="One year of flows, stylized (base 100): new loans create deposits, repayments destroy them.",
        note="Repaying principal extinguishes the deposit that faces it; interest is the bank's income —\n"
             "it changes hands without being destroyed. The money stock follows the net balance."),
}

def build_figure(content: tuple, lang: str) -> Figure:
    """Trois barres horizontales (brut, remboursé, net) puis deux cartes d'épisodes."""
    text = LABELS[lang]
    bars, cases = content
    fig = nm.figure(height_px=1120)
    ax = nm.blank_axes(fig)

    # Barres de flux.
    x_left, x_max, vmax = 480, 1690, 100
    bar_h, centers = 86, [810, 668, 526]
    for bar, cy in zip(bars, centers):
        length = bar["value"] / vmax * (x_max - x_left)
        color = nm.COLORS[bar["color"]]
        nm.card(ax, x_left, cy - bar_h / 2, length, bar_h, edge=color, fill=color, radius=14)
        ax.text(68, cy, bar["name"], ha="left", va="center", fontsize=27,
                fontweight="bold", color=color)
        if length > 220:
            ax.text(x_left + length - 26, cy, bar["label"], ha="right", va="center",
                    fontsize=30, fontweight="bold", color=nm.COLORS["bg"])
        else:
            ax.text(x_left + length + 26, cy, bar["label"], ha="left", va="center",
                    fontsize=30, fontweight="bold", color=color)

    # Deux épisodes historiques.
    for i, (title, body) in enumerate(cases):
        x = 65 + i * 837
        nm.card(ax, x, 165, 780, 270, edge=nm.COLORS["edge"], lw=2.2, radius=22)
        ax.text(x + 42, 387, title, ha="left", va="center", fontsize=26,
                fontweight="bold", color=nm.COLORS["text"])
        ax.text(x + 42, 329, body, ha="left", va="top", fontsize=22.5,
                color=nm.COLORS["muted"], linespacing=1.45)

    nm.header(fig, text["title"], text["sub"])
    nm.footer(fig, text["note"])
    return fig

build_figure(flows(LANG), LANG)
```
