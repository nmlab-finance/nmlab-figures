```python
LANG = "fr"   # "fr" ou "en" — langue des libellés / label language

# Récupère puis active le style partagé NMLab (thème sombre + police Inter).
# Fetch and activate the shared NMLab style (dark theme + Inter font).
import urllib.request

urllib.request.urlretrieve("https://raw.githubusercontent.com/nmlab-finance/nmlab-figures/main/nmlab_style.py", "nmlab_style.py")
import nmlab_style as nm

nm.setup()


def brakes(lang: str) -> list[dict]:
    """Les quatre freins (titre, corps, couleur), localisés.
    The four brakes (title, body, color), localized."""
    if lang == "fr":
        return [
            dict(title="1 · La demande solvable", color="blue",
                 body="Un crédit n'existe que si un emprunteur\nsolvable le demande — et si la marge\ncouvre le risque, les coûts et le\ncapital mobilisé."),
            dict(title="2 · Les fonds propres (Bâle III)", color="rose",
                 body="Chaque prêt doit être adossé à du\ncapital : CET1 ≥ 4,5 % des actifs\npondérés, coussins en plus — les grandes\nbanques de l'UE : ≈ 16 % fin 2024."),
            dict(title="3 · La liquidité et les fuites", color="amber",
                 body="Espèces retirées, paiements vers d'autres\nbanques, réserves obligatoires (1 % en\nzone euro) : prêter oblige à se\nrefinancer — LCR et NSFR l'encadrent."),
            dict(title="4 · Le prix de la banque centrale", color="teal",
                 body="Le taux directeur fixe le coût de la\nmonnaie centrale ; il se répercute sur\nles taux débiteurs et freine ou stimule\nla demande de crédit."),
        ]
    return [
        dict(title="1 · Creditworthy demand", color="blue",
             body="A loan only exists if a solvent\nborrower asks for it — and if the margin\ncovers the risk, the costs and the\ncapital tied up."),
        dict(title="2 · Capital (Basel III)", color="rose",
             body="Every loan must be backed by capital:\nCET1 ≥ 4.5% of risk-weighted assets,\nbuffers on top — large EU banks\nreport ≈ 16% in late 2024."),
        dict(title="3 · Liquidity and leakages", color="amber",
             body="Cash withdrawals, payments to other\nbanks, minimum reserves (1% in the euro\narea): lending forces refinancing —\nLCR and NSFR frame it."),
        dict(title="4 · The central bank's price", color="teal",
             body="The policy rate sets the cost of\ncentral-bank money; it passes through\nto lending rates and curbs or spurs\ncredit demand."),
    ]


from matplotlib.figure import Figure

LABELS = {
    "fr": dict(
        title="Pourquoi les banques ne prêtent pas sans limite",
        sub="Quatre freins bornent la création de crédit : la rentabilité, le capital, la liquidité et la politique monétaire.",
        note="Le frein n'est pas un stock de dépôts à prêter : c'est un ensemble de prix, de règles et de risques qui rend\n"
             "chaque crédit supplémentaire plus ou moins attrayant."),
    "en": dict(
        title="Why banks do not lend without limit",
        sub="Four brakes bound credit creation: profitability, capital, liquidity and monetary policy.",
        note="The brake is not a stock of deposits waiting to be lent: it is a set of prices, rules and risks that makes\n"
             "each additional loan more or less attractive."),
}

def build_figure(cards: list[dict], lang: str) -> Figure:
    """Grille 2 × 2 de cartes, une par frein."""
    text = LABELS[lang]
    fig = nm.figure(height_px=1080)
    ax = nm.blank_axes(fig)

    xs, ys = [65, 902], [480, 120]
    for i, c in enumerate(cards):
        x, y = xs[i % 2], ys[i // 2]
        color = nm.COLORS[c["color"]]
        nm.card(ax, x, y, 780, 340, edge=color, lw=2.4, radius=22)
        ax.text(x + 42, y + 284, c["title"], ha="left", va="center", fontsize=27,
                fontweight="bold", color=color)
        ax.text(x + 42, y + 206, c["body"], ha="left", va="top", fontsize=22.5,
                color=nm.COLORS["muted"], linespacing=1.42)

    nm.header(fig, text["title"], text["sub"])
    nm.footer(fig, text["note"])
    return fig

build_figure(brakes(LANG), LANG)
```
