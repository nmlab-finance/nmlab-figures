```python
LANG = "fr"   # "fr" ou "en" — langue des libellés / label language

# Récupère puis active le style partagé NMLab (thème sombre + police Inter).
# Fetch and activate the shared NMLab style (dark theme + Inter font).
import urllib.request

urllib.request.urlretrieve("https://raw.githubusercontent.com/nmlab-finance/nmlab-figures/main/nmlab_style.py", "nmlab_style.py")
import nmlab_style as nm

nm.setup()


def taps(lang: str) -> list[tuple[str, str, str]]:
    """Les trois robinets du tableau de bord : (titre, description, sources).
    The three taps on the dashboard: (title, description, sources)."""
    if lang == "fr":
        return [
            ("Le bilan de la\nbanque centrale",
             "Hebdomadaire. QE ou QT :\nla taille donne le sens\ndu robinet officiel.",
             "BCE · Fed (H.4.1)"),
            ("Le flux\nde crédit",
             "Mensuel. La porte principale\nde la création — les cadrans\ndu chapitre 22.",
             "BCE · Webstat"),
            ("Le budget\nde l'État",
             "Déficit et calendrier\nd'émissions : la porte\nbudgétaire de la monnaie.",
             "AFT · Treasury"),
        ]
    return [
        ("The central bank's\nbalance sheet",
         "Weekly. QE or QT: its size\nsets the direction of the\nofficial tap.",
         "ECB · Fed (H.4.1)"),
        ("The flow\nof credit",
         "Monthly. The main door of\ncreation — chapter 22's\ndials.",
         "ECB · Webstat"),
        ("The state's\nbudget",
         "Deficit and issuance\ncalendar: money's fiscal\ndoor.",
         "AFT · Treasury"),
    ]


from matplotlib.figure import Figure

LABELS = {
    "fr": dict(
        title="Trois robinets à surveiller",
        sub="La création monétaire se lit en trois publications régulières — puis une seule question la résume.",
        question="La question qui résume tout : la monnaie nouvelle atteint-elle les comptes du\n"
                 "public — ou reste-t-elle dans le circuit fermé des réserves ?",
        note="Le chapitre 29 refermera le module : comment relier ces robinets aux prix des actifs sans confondre\n"
             "corrélation et mécanique."),
    "en": dict(
        title="Three taps to watch",
        sub="Money creation can be read in three regular releases — then a single question sums it up.",
        question="The question that sums it all up: does the new money reach the public's\n"
                 "accounts — or does it stay in the closed circuit of reserves?",
        note="Chapter 29 will close the module: how to tie these taps to asset prices without mistaking\n"
             "correlation for mechanics."),
}

def build_figure(taps_: list[tuple[str, str, str]], lang: str) -> Figure:
    """Trois cartes « robinet » (titre, description, sources) puis la carte-question."""
    text = LABELS[lang]
    fig = nm.figure(height_px=1040)
    ax = nm.blank_axes(fig)

    card_w, gap, x0 = 515, 39, 62
    top, bottom = 780, 360
    for i, (t, d, src) in enumerate(taps_):
        x = x0 + i * (card_w + gap)
        nm.card(ax, x, bottom, card_w, top - bottom, edge=nm.COLORS["blue"], lw=2.4, radius=22)
        ax.text(x + 40, top - 44, t, ha="left", va="top", fontsize=25,
                fontweight="bold", color=nm.COLORS["text"], linespacing=1.35)
        ax.text(x + 40, top - 168, d, ha="left", va="top", fontsize=21,
                color=nm.COLORS["muted"], linespacing=1.5)
        ax.text(x + 40, bottom + 42, src, ha="left", va="center", fontsize=19.5,
                fontweight="bold", color=nm.COLORS["blue"])

    nm.card(ax, 62, 150, 1623, 158, edge=nm.COLORS["amber"], lw=2.4, radius=22)
    ax.text(873, 229, text["question"], ha="center", va="center", fontsize=23.5,
            fontweight="bold", color=nm.COLORS["amber"], linespacing=1.5)

    nm.header(fig, text["title"], text["sub"])
    nm.footer(fig, text["note"])
    return fig

build_figure(taps(LANG), LANG)
```
