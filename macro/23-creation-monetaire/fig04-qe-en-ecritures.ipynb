```python
LANG = "fr"   # "fr" ou "en" — langue des libellés / label language

# Récupère puis active le style partagé NMLab (thème sombre + police Inter).
# Fetch and activate the shared NMLab style (dark theme + Inter font).
import urllib.request

urllib.request.urlretrieve("https://raw.githubusercontent.com/nmlab-finance/nmlab-figures/main/nmlab_style.py", "nmlab_style.py")
import nmlab_style as nm

nm.setup()


def qe_accounts(lang: str) -> list[dict]:
    """Trois comptes en T stylisés pour 100 d'achats de titres via un fonds.
    Three stylized T-accounts for 100 of asset purchases via a fund."""
    if lang == "fr":
        return [
            dict(who="Banque centrale",
                 actif=["+100 obligation"], passif=["+100 réserves"]),
            dict(who="Banque du fonds",
                 actif=["+100 réserves"], passif=["+100 dépôt\ndu fonds"]),
            dict(who="Fonds de pension",
                 actif=["−100 obligation", "+100 dépôt"], passif=["—"]),
        ]
    return [
        dict(who="Central bank",
             actif=["+100 bond"], passif=["+100 reserves"]),
        dict(who="The fund's bank",
             actif=["+100 reserves"], passif=["+100 fund\ndeposit"]),
        dict(who="Pension fund",
             actif=["−100 bond", "+100 deposit"], passif=["—"]),
    ]


from matplotlib.figure import Figure

LABELS = {
    "fr": dict(
        title="100 € de QE : qui reçoit quoi ?",
        sub="La banque centrale achète une obligation à un fonds de pension — trois bilans bougent d'un coup.",
        actif="Actif", passif="Passif",
        variant="Variante : si le vendeur est une banque — réserves +100, dépôt 0. Rien n'atteint le public.",
        note="Écritures stylisées d'après Bank of England (2014) : acheté à un non-bancaire, le titre est payé en\n"
             "dépôt neuf, et la banque du vendeur reçoit les réserves correspondantes — deux monnaies d'un geste."),
    "en": dict(
        title="100 of QE: who receives what?",
        sub="The central bank buys a bond from a pension fund — three balance sheets move at once.",
        actif="Assets", passif="Liabilities",
        variant="Variant: if the seller is a bank — reserves +100, deposits 0. Nothing reaches the public.",
        note="Stylized entries after the Bank of England (2014): bought from a non-bank, the bond is paid for with a\n"
             "brand-new deposit, and the seller's bank receives the matching reserves — two monies in one stroke."),
}

def build_figure(accounts: list[dict], lang: str) -> Figure:
    """Trois comptes en T côte à côte, puis une carte « variante » pleine largeur."""
    text = LABELS[lang]
    fig = nm.figure(height_px=1050)
    ax = nm.blank_axes(fig)

    card_w, gap, x0 = 515, 39, 62
    top, bottom = 810, 440
    for i, acc in enumerate(accounts):
        x = x0 + i * (card_w + gap)
        nm.card(ax, x, bottom, card_w, top - bottom, edge=nm.COLORS["edge"], lw=2.2, radius=22)
        ax.text(x + card_w / 2, top - 54, acc["who"], ha="center", va="center",
                fontsize=27, fontweight="bold", color=nm.COLORS["text"])
        mid = x + card_w / 2
        ax.plot([x + 30, x + card_w - 30], [top - 100, top - 100],
                color=nm.COLORS["edge"], lw=2)
        ax.plot([mid, mid], [top - 100, bottom + 36], color=nm.COLORS["edge"], lw=2)
        ax.text((x + 30 + mid) / 2, top - 138, text["actif"], ha="center", va="center",
                fontsize=20, fontstyle="italic", color=nm.COLORS["muted"])
        ax.text((mid + x + card_w - 30) / 2, top - 138, text["passif"], ha="center",
                va="center", fontsize=20, fontstyle="italic", color=nm.COLORS["muted"])
        for j, line in enumerate(acc["actif"]):
            color = nm.COLORS["rose"] if line.startswith("−") else nm.COLORS["green"]
            ax.text((x + 30 + mid) / 2, top - 215 - j * 74, line, ha="center", va="center",
                    fontsize=19, fontweight="bold", color=color, linespacing=1.4)
        for j, line in enumerate(acc["passif"]):
            color = nm.COLORS["muted"] if line == "—" else nm.COLORS["green"]
            ax.text((mid + x + card_w - 30) / 2, top - 215 - j * 74, line, ha="center",
                    va="center", fontsize=19, fontweight="bold", color=color, linespacing=1.4)

    nm.card(ax, 62, 240, 1623, 120, edge=nm.COLORS["amber"], lw=2.4, radius=22)
    ax.text(873, 300, text["variant"], ha="center", va="center", fontsize=23.5,
            fontweight="bold", color=nm.COLORS["amber"])

    nm.header(fig, text["title"], text["sub"])
    nm.footer(fig, text["note"])
    return fig

build_figure(qe_accounts(LANG), LANG)
```
