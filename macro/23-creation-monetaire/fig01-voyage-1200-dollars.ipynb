```python
LANG = "fr"   # "fr" ou "en" — langue des libellés / label language

# Récupère puis active le style partagé NMLab (thème sombre + police Inter).
# Fetch and activate the shared NMLab style (dark theme + Inter font).
import urllib.request

urllib.request.urlretrieve("https://raw.githubusercontent.com/nmlab-finance/nmlab-figures/main/nmlab_style.py", "nmlab_style.py")
import nmlab_style as nm

nm.setup()


def journey_cards(lang: str) -> list[tuple[str, str]]:
    """Les trois étapes du paiement fédéral : (titre, corps), localisées.
    The three steps of the federal payment: (title, body), localized."""
    if lang == "fr":
        return [
            ("Le Trésor américain",
             "Ordonne le paiement\ndepuis son compte à la\nFed (le TGA) : le solde\nde l'État baisse."),
            ("La banque du ménage",
             "Reçoit 1 200 dollars\nde réserves sur son\ncompte à la Fed —\nmonnaie interbancaire."),
            ("Le compte du ménage",
             "La banque crédite un\ndépôt neuf : la monnaie\ndu public vient\nd'apparaître."),
        ]
    return [
        ("The U.S. Treasury",
         "Orders the payment out\nof its account at the\nFed (the TGA): the\nstate's balance falls."),
        ("The household's bank",
         "Receives 1,200 dollars\nof reserves in its\naccount at the Fed —\ninterbank money."),
        ("Household's account",
         "The bank credits a\nbrand-new deposit: the\npublic's money has\njust appeared."),
    ]


from matplotlib.figure import Figure

LABELS = {
    "fr": dict(
        title="Avril 2020 : le voyage des 1 200 dollars",
        sub="Le CARES Act paie les ménages : le Trésor dépense, la banque reçoit des réserves — et crée le dépôt.",
        pill="CARES Act, 27 mars 2020 — premiers virements à la mi-avril",
        arrow1="réserves", arrow2="dépôt",
        note="Aucun épargnant n'est débité en face : le dépôt du ménage est créé en route. Deux monnaies avancent\n"
             "de concert — les réserves à l'étage interbancaire, le dépôt à l'étage du public."),
    "en": dict(
        title="April 2020: the journey of the 1,200 dollars",
        sub="The CARES Act pays households: the Treasury spends, the bank receives reserves — and creates the deposit.",
        pill="CARES Act, March 27, 2020 — first transfers in mid-April",
        arrow1="reserves", arrow2="deposit",
        note="No saver is debited on the other side: the household's deposit is created along the way. Two monies move\n"
             "in step — reserves on the interbank floor, the deposit on the public's floor."),
}

def build_figure(cards: list[tuple[str, str]], lang: str) -> Figure:
    """Pastille de contexte puis trois cartes reliées par deux flèches légendées."""
    text = LABELS[lang]
    fig = nm.figure(height_px=1040)
    ax = nm.blank_axes(fig)

    # Pastille de contexte, au-dessus des cartes.
    nm.card(ax, 353, 680, 1040, 74, edge=nm.COLORS["blue"], radius=37, lw=2.2)
    ax.text(873, 717, text["pill"], ha="center", va="center", fontsize=23,
            fontweight="bold", color=nm.COLORS["blue"])

    # Trois cartes, séparées par des couloirs assez larges pour les étiquettes.
    card_w, gap, x0 = 445, 135, 61
    top, bottom = 600, 255
    centers = []
    for i, (title, body) in enumerate(cards):
        x = x0 + i * (card_w + gap)
        centers.append((x, x + card_w))
        nm.card(ax, x, bottom, card_w, top - bottom, edge=nm.COLORS["edge"], lw=2.2, radius=22)
        ax.text(x + 38, top - 58, title, ha="left", va="center", fontsize=25.5,
                fontweight="bold", color=nm.COLORS["text"])
        ax.text(x + 38, top - 128, body, ha="left", va="top", fontsize=21,
                color=nm.COLORS["muted"], linespacing=1.5)

    # Flèches entre cartes, étiquette centrée dans le couloir, sous la flèche.
    y_mid = (top + bottom) / 2 + 20
    for i, label in ((0, text["arrow1"]), (1, text["arrow2"])):
        x_from, x_to = centers[i][1] + 10, centers[i + 1][0] - 10
        ax.annotate("", xy=(x_to, y_mid), xytext=(x_from, y_mid),
                    arrowprops=dict(arrowstyle="-|>", lw=3, color=nm.COLORS["amber"],
                                    mutation_scale=26))
        ax.text((x_from + x_to) / 2, y_mid - 26, label, ha="center", va="top",
                fontsize=18.5, fontstyle="italic", color=nm.COLORS["amber"])

    nm.header(fig, text["title"], text["sub"])
    nm.footer(fig, text["note"])
    return fig

build_figure(journey_cards(LANG), LANG)
```
