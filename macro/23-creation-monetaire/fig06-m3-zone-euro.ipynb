```python
LANG = "fr"   # "fr" ou "en" — langue des libellés / label language

# Récupère puis active le style partagé NMLab (thème sombre + police Inter).
# Fetch and activate the shared NMLab style (dark theme + Inter font).
import urllib.request

urllib.request.urlretrieve("https://raw.githubusercontent.com/nmlab-finance/nmlab-figures/main/nmlab_style.py", "nmlab_style.py")
import nmlab_style as nm

nm.setup()


from pandas import Series


def load_m3() -> Series:
    """Croissance annuelle de M3 en zone euro, %, mensuelle (BCE, série BSI,
    relevée le 24 juillet 2026 — points embarqués pour rester exécutable hors ligne).
    Euro-area M3 annual growth, %, monthly (ECB BSI series, embedded points)."""
    import pandas as pd

    M3_YOY = [  # croissance annuelle de M3, %, mensuel de 2013-01 a 2026-05 (BCE, BSI)
        # 2013
        3.42, 3.08, 2.40, 3.07, 2.80, 2.28, 1.99, 2.27, 2.03, 1.50, 1.50, 1.00,
        # 2014
        1.15, 1.31, 1.12, 0.74, 1.18, 1.71, 1.88, 1.95, 2.37, 2.51, 3.04, 3.74,
        # 2015
        3.70, 4.00, 4.31, 5.21, 4.70, 4.64, 4.93, 4.80, 4.63, 5.15, 4.97, 4.72,
        # 2016
        5.07, 5.06, 5.05, 4.68, 4.86, 5.04, 4.86, 4.42, 4.98, 4.43, 4.76, 5.00,
        # 2017
        4.86, 4.72, 5.22, 4.86, 4.96, 4.99, 4.51, 5.20, 5.20, 4.99, 4.94, 4.58,
        # 2018
        4.55, 4.30, 3.72, 3.87, 3.99, 4.42, 4.03, 3.52, 3.56, 3.86, 3.74, 4.08,
        # 2019
        3.83, 4.28, 4.60, 4.71, 4.80, 4.46, 5.13, 5.65, 5.55, 5.60, 5.61, 4.95,
        # 2020
        5.19, 5.46, 7.44, 8.19, 8.85, 9.15, 10.09, 9.47, 10.35, 10.51, 10.98, 12.23,
        # 2021
        12.50, 12.36, 10.04, 9.19, 8.44, 8.27, 7.61, 7.87, 7.40, 7.68, 7.63, 6.94,
        # 2022
        6.83, 6.55, 6.55, 6.19, 5.79, 5.68, 5.53, 6.09, 6.21, 5.10, 4.79, 3.98,
        # 2023
        3.53, 2.94, 2.51, 1.87, 1.28, 0.63, -0.36, -1.29, -1.18, -0.96, -0.93, 0.05,
        # 2024
        0.13, 0.40, 0.84, 1.39, 1.49, 2.21, 2.34, 2.90, 3.21, 3.40, 3.79, 3.63,
        # 2025
        3.66, 3.88, 3.69, 3.89, 3.93, 3.43, 3.31, 2.92, 2.78, 2.78, 3.03, 2.79,
        # 2026
        3.18, 2.94, 3.23, 2.72, 3.20,
    ]
    idx = pd.date_range("2013-01-01", periods=len(M3_YOY), freq="MS")
    return pd.Series(M3_YOY, index=idx, name="M3 yoy")


from matplotlib.figure import Figure
from pandas import Series

LABELS = {
    "fr": dict(
        title="La masse monétaire respire : M3 en zone euro",
        sub="Croissance annuelle de M3, en % — quand les portes créent plus qu'elles ne détruisent, la courbe monte.",
        qe="mars 2015 :\nle QE démarre",
        peak="+12,5 % — janv. 2021",
        trough="−1,3 % — août 2023,\npremière contraction depuis 2010",
        last="+3,2 %\nmai 2026",
        note="Taux de croissance annuel de l'agrégat M3 (indice des encours notionnels). Source : BCE (statistiques\n"
             "BSI), relevé du 24 juillet 2026."),
    "en": dict(
        title="The money stock breathes: euro-area M3",
        sub="Annual M3 growth, % — when the doors create more than they destroy, the line climbs.",
        qe="March 2015:\nQE starts",
        peak="+12.5% — Jan. 2021",
        trough="−1.3% — Aug. 2023,\nfirst contraction since 2010",
        last="+3.2%\nMay 2026",
        note="Annual growth rate of the M3 aggregate (index of notional stocks). Source: ECB (BSI statistics),\n"
             "retrieved July 24, 2026."),
}

def build_figure(m3: Series, lang: str) -> Figure:
    """Courbe M3 avec ligne de zéro, jalon QE 2015 et annotations pic / creux / dernier point."""
    import pandas as pd

    text = LABELS[lang]
    fig = nm.figure(height_px=1045)
    ax = nm.axes(fig, left=0.056)
    ax.plot(m3.index, m3, color=nm.COLORS["blue"], lw=3.2, zorder=3)
    ax.axhline(0, color=nm.COLORS["muted"], lw=1.6, alpha=0.85, zorder=2)
    ax.set_ylim(-3.6, 14.4)
    ax.set_xlim(pd.Timestamp("2012-08-01"), pd.Timestamp("2028-09-01"))

    qe = pd.Timestamp("2015-03-01")
    ax.axvline(qe, color=nm.COLORS["edge"], lw=2, linestyle=(0, (5, 5)), zorder=1)
    ax.text(qe, 13.9, text["qe"], ha="left", va="top", fontsize=19,
            color=nm.COLORS["muted"], linespacing=1.4, transform=ax.transData)

    peak = m3.idxmax()
    ax.scatter([peak], [m3.max()], s=90, color=nm.COLORS["amber"], zorder=4)
    ax.text(peak, m3.max() + 0.75, text["peak"], ha="center", va="bottom",
            fontsize=21, fontweight="bold", color=nm.COLORS["amber"])

    trough = m3.idxmin()
    ax.scatter([trough], [m3.min()], s=90, color=nm.COLORS["rose"], zorder=4)
    ax.text(trough - pd.Timedelta(days=75), m3.min() - 0.55, text["trough"],
            ha="right", va="top", fontsize=19.5, color=nm.COLORS["rose"], linespacing=1.45)

    ax.text(m3.index[-1] + pd.Timedelta(days=40), m3.iloc[-1], text["last"],
            ha="left", va="center", fontsize=19.5, fontweight="bold",
            color=nm.COLORS["blue"], linespacing=1.35)

    nm.header(fig, text["title"], text["sub"])
    nm.footer(fig, text["note"])
    return fig

build_figure(load_m3(), LANG)
```
