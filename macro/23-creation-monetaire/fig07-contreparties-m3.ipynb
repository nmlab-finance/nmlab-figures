```python
LANG = "fr"   # "fr" ou "en" — langue des libellés / label language

# Récupère puis active le style partagé NMLab (thème sombre + police Inter).
# Fetch and activate the shared NMLab style (dark theme + Inter font).
import urllib.request

urllib.request.urlretrieve("https://raw.githubusercontent.com/nmlab-finance/nmlab-figures/main/nmlab_style.py", "nmlab_style.py")
import nmlab_style as nm

nm.setup()


def episodes(lang: str) -> list[dict]:
    """Deux épisodes lus par les contreparties de M3 : lignes (libellé, tendance).
    Two episodes read through the counterparts of M3: rows (label, trend)."""
    if lang == "fr":
        return [
            dict(year="2020-2021", verdict="M3 : +12,5 %", color="blue", up=True, rows=[
                ("Crédit au secteur privé", "hausse"),
                ("Crédit aux administrations", "forte hausse"),
                ("Épargne longue (hors M3)", "faible"),
            ]),
            dict(year="2023", verdict="M3 : −1,3 %", color="rose", up=False, rows=[
                ("Crédit au secteur privé", "à l'arrêt"),
                ("Crédit aux administrations", "baisse (QT)"),
                ("Épargne longue (hors M3)", "aspire les dépôts"),
            ]),
        ]
    return [
        dict(year="2020-2021", verdict="M3: +12.5%", color="blue", up=True, rows=[
            ("Credit to the private sector", "rising"),
            ("Credit to governments", "surging"),
            ("Long-term savings", "weak"),
        ]),
        dict(year="2023", verdict="M3: −1.3%", color="rose", up=False, rows=[
            ("Credit to the private sector", "stalled"),
            ("Credit to governments", "falling (QT)"),
            ("Long-term savings", "drains deposits"),
        ]),
    ]


from matplotlib.figure import Figure

LABELS = {
    "fr": dict(
        title="Les contreparties de M3 : d'où venait la monnaie",
        sub="La comptabilité de la BCE relie chaque variation de M3 à ses sources — deux épisodes en regard.",
        note="Lecture des « contreparties » : crédit aux résidents (privé + administrations) + créances nettes sur\n"
             "l'extérieur − engagements financiers à long terme. Source : BCE, statistiques BSI."),
    "en": dict(
        title="The counterparts of M3: where the money came from",
        sub="The ECB's accounting ties every move in M3 back to its sources — two episodes side by side.",
        note="Reading the \"counterparts\": credit to residents (private + governments) + net external assets −\n"
             "longer-term financial liabilities. Source: ECB, BSI statistics."),
}

def build_figure(episodes_: list[dict], lang: str) -> Figure:
    """Deux panneaux d'épisode : trois lignes de contreparties et le verdict M3."""
    text = LABELS[lang]
    fig = nm.figure(height_px=1150)
    ax = nm.blank_axes(fig)

    panel_w, x0, gap = 790, 62, 43
    top, bottom = 930, 210
    for i, ep in enumerate(episodes_):
        x = x0 + i * (panel_w + gap)
        nm.card(ax, x, bottom, panel_w, top - bottom, edge=nm.COLORS[ep["color"]], lw=2.6, radius=24)
        ax.text(x + 42, top - 62, ep["year"], ha="left", va="center", fontsize=30,
                fontweight="bold", color=nm.COLORS[ep["color"]])
        for j, (label, trend) in enumerate(ep["rows"]):
            y = top - 150 - j * 128
            nm.card(ax, x + 42, y - 96, panel_w - 84, 100, edge=nm.COLORS["edge"],
                    fill=nm.COLORS["bg"], lw=2, radius=18)
            ax.text(x + 74, y - 46, label, ha="left", va="center", fontsize=21.5,
                    color=nm.COLORS["text"])
            ax.text(x + panel_w - 74, y - 46, trend, ha="right", va="center",
                    fontsize=21.5, fontweight="bold", color=nm.COLORS[ep["color"]])
        nm.card(ax, x + 42, bottom + 40, panel_w - 84, 88,
                edge=nm.COLORS[ep["color"]], lw=2.4, radius=30)
        ax.text(x + panel_w / 2, bottom + 84, ep["verdict"], ha="center", va="center",
                fontsize=26, fontweight="bold", color=nm.COLORS[ep["color"]])

    nm.header(fig, text["title"], text["sub"])
    nm.footer(fig, text["note"])
    return fig

build_figure(episodes(LANG), LANG)
```
