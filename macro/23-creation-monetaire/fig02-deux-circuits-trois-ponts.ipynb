```python
LANG = "fr"   # "fr" ou "en" — langue des libellés / label language

# Récupère puis active le style partagé NMLab (thème sombre + police Inter).
# Fetch and activate the shared NMLab style (dark theme + Inter font).
import urllib.request

urllib.request.urlretrieve("https://raw.githubusercontent.com/nmlab-finance/nmlab-figures/main/nmlab_style.py", "nmlab_style.py")
import nmlab_style as nm

nm.setup()


def floors(lang: str) -> list[dict]:
    """Les deux étages de la monnaie : titre, occupants, usage — localisés.
    The two floors of money: title, occupants, use — localized."""
    if lang == "fr":
        return [
            dict(title="L'étage central — les réserves",
                 who=["Banques", "Trésor public", "Banque centrale"],
                 use="Règlements entre banques (T2),\npolitique monétaire.",
                 color="blue"),
            dict(title="L'étage du public — dépôts et billets",
                 who=["Ménages", "Entreprises", "Fonds, assureurs"],
                 use="Paiements de tous les jours,\népargne liquide (M3).",
                 color="teal"),
        ]
    return [
        dict(title="The central floor — reserves",
             who=["Banks", "The Treasury", "Central bank"],
             use="Settlement between banks,\nmonetary policy.",
             color="blue"),
        dict(title="The public's floor — deposits and cash",
             who=["Households", "Firms", "Funds, insurers"],
             use="Everyday payments,\nliquid savings (M3).",
             color="teal"),
    ]

def bridges(lang: str) -> list[str]:
    """Les trois ponts qui relient les deux étages. The three bridges between floors."""
    if lang == "fr":
        return ["Billets\n(retraits)", "Dépenses et impôts\nde l'État", "Achats et ventes de\nla banque centrale"]
    return ["Banknotes\n(withdrawals)", "Government spending\nand taxes", "Central-bank asset\npurchases and sales"]


from matplotlib.figure import Figure

LABELS = {
    "fr": dict(
        title="Deux monnaies, deux circuits — et trois ponts",
        sub="Les réserves circulent entre comptes à la banque centrale ; les dépôts, entre comptes du public.",
        note="Les réserves ne quittent jamais l'étage central : le public n'y a pas de compte. Les ponts font naître\n"
             "ou disparaître la monnaie d'un étage quand l'autre bouge — c'est là que ce chapitre se joue."),
    "en": dict(
        title="Two monies, two circuits — and three bridges",
        sub="Reserves move between accounts at the central bank; deposits, between the public's accounts.",
        note="Reserves never leave the central floor: the public holds no account there. The bridges make one floor's\n"
             "money appear or vanish when the other moves — this chapter lives on those bridges."),
}

def build_figure(floors_: list[dict], bridges_: list[str], lang: str) -> Figure:
    """Deux bandes pleine largeur (étages) reliées par trois flèches doubles légendées."""
    text = LABELS[lang]
    fig = nm.figure(height_px=1120)
    ax = nm.blank_axes(fig)

    bands = [(628, 900), (140, 412)]                     # (bas, haut) de chaque bande
    for spec, (b, t) in zip(floors_, bands):
        nm.card(ax, 62, b, 1623, t - b, edge=nm.COLORS[spec["color"]], lw=2.6, radius=24)
        ax.text(108, t - 58, spec["title"], ha="left", va="center", fontsize=29,
                fontweight="bold", color=nm.COLORS[spec["color"]])
        for i, who in enumerate(spec["who"]):
            x = 108 + i * 330
            nm.card(ax, x, b + 52, 300, 72, edge=nm.COLORS["edge"],
                    fill=nm.COLORS["bg"], lw=2, radius=18)
            ax.text(x + 150, b + 88, who, ha="center", va="center", fontsize=22,
                    color=nm.COLORS["text"])
        ax.text(1640, (b + t) / 2 - 12, spec["use"], ha="right", va="center",
                fontsize=21, color=nm.COLORS["muted"], linespacing=1.5)

    # Trois ponts : flèches doubles entre les bandes, étiquette à droite de chaque flèche.
    xs = (300, 800, 1300)
    for x, label in zip(xs, bridges_):
        ax.annotate("", xy=(x, bands[0][0] - 8), xytext=(x, bands[1][1] + 8),
                    arrowprops=dict(arrowstyle="<|-|>", lw=3, color=nm.COLORS["amber"],
                                    mutation_scale=24))
        ax.text(x + 26, (bands[0][0] + bands[1][1]) / 2, label, ha="left", va="center",
                fontsize=20.5, fontstyle="italic", color=nm.COLORS["amber"], linespacing=1.4)

    nm.header(fig, text["title"], text["sub"])
    nm.footer(fig, text["note"])
    return fig

build_figure(floors(LANG), bridges(LANG), LANG)
```
