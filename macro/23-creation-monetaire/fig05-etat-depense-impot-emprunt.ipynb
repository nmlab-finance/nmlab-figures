```python
LANG = "fr"   # "fr" ou "en" — langue des libellés / label language

# Récupère puis active le style partagé NMLab (thème sombre + police Inter).
# Fetch and activate the shared NMLab style (dark theme + Inter font).
import urllib.request

urllib.request.urlretrieve("https://raw.githubusercontent.com/nmlab-finance/nmlab-figures/main/nmlab_style.py", "nmlab_style.py")
import nmlab_style as nm

nm.setup()


def state_flows(lang: str) -> tuple[list[tuple[str, str]], list[tuple[str, str, str]]]:
    """Circuit quotidien de l'État et destinataires de sa dette, localisés.
    The state's daily circuit and who absorbs its debt, localized."""
    if lang == "fr":
        daily = [
            ("La dépense publique", "le compte du Trésor se vide —\nréserves et dépôts montent"),
            ("L'impôt", "le chemin inverse : le dépôt\ns'éteint, le Trésor se remplit"),
            ("Le solde de l'année", "ce qui ne se compense pas\nest le déficit — à financer"),
        ]
        buyers = [
            ("Une banque", "achète l'obligation en créant\nun dépôt : monnaie nouvelle", "blue"),
            ("Un ménage, un fonds", "paie avec un dépôt existant :\nsimple transfert", "teal"),
            ("La banque centrale, en direct", "interdit — article 123 du TFUE ;\nmarché secondaire seulement", "rose"),
        ]
    else:
        daily = [
            ("Public spending", "the Treasury's account drains —\nreserves and deposits rise"),
            ("Taxes", "the reverse path: the deposit\ndies, the Treasury refills"),
            ("The year's balance", "what does not net out is\nthe deficit — to be funded"),
        ]
        buyers = [
            ("A bank", "buys the bond by creating\na deposit: new money", "blue"),
            ("A household, a fund", "pays with an existing deposit:\na mere transfer", "teal"),
            ("The central bank, directly", "forbidden — Article 123 TFEU;\nsecondary market only", "rose"),
        ]
    return daily, buyers


from matplotlib.figure import Figure

LABELS = {
    "fr": dict(
        title="L'État : la dépense, l'impôt, l'emprunt",
        sub="Chaque euro public passe par le compte du Trésor à la banque centrale — puis le déficit cherche preneur.",
        left_head="Le circuit quotidien",
        right_head="Le déficit s'emprunte : qui achète ?",
        note="Un déficit ne crée de la monnaie que selon la porte par laquelle il se finance — d'où l'éternel débat\n"
             "sur la « monétisation ». L'interdit de l'article 123 ferme la porte directe, pas le marché secondaire."),
    "en": dict(
        title="The state: spending, taxing, borrowing",
        sub="Every public euro flows through the Treasury's account at the central bank — then the deficit seeks takers.",
        left_head="The daily circuit",
        right_head="The deficit is borrowed: who buys?",
        note="A deficit creates money only through the door that finances it — hence the eternal \"monetization\"\n"
             "debate. Article 123 bolts the direct door, not the secondary market."),
}

def build_figure(flows: tuple, lang: str) -> Figure:
    """Colonne gauche : dépense et impôt ; colonne droite : les trois acheteurs de dette."""
    text = LABELS[lang]
    fig = nm.figure(height_px=1150)
    ax = nm.blank_axes(fig)

    daily, buyers = flows
    ax.text(68, 952, text["left_head"], ha="left", va="center", fontsize=26,
            fontweight="bold", color=nm.COLORS["text"])
    for i, (t, d) in enumerate(daily):
        y_top = 880 - i * 260
        nm.card(ax, 62, y_top - 232, 760, 232, edge=nm.COLORS["edge"], lw=2.2, radius=22)
        ax.text(100, y_top - 56, t, ha="left", va="center", fontsize=26,
                fontweight="bold", color=nm.COLORS["text"])
        ax.text(100, y_top - 118, d, ha="left", va="top", fontsize=21.5,
                color=nm.COLORS["muted"], linespacing=1.5)

    ax.text(930, 952, text["right_head"], ha="left", va="center", fontsize=26,
            fontweight="bold", color=nm.COLORS["text"])
    for i, (t, d, color) in enumerate(buyers):
        y_top = 880 - i * 260
        nm.card(ax, 924, y_top - 232, 760, 232, edge=nm.COLORS[color], lw=2.4, radius=22)
        ax.text(962, y_top - 56, t, ha="left", va="center", fontsize=26,
                fontweight="bold", color=nm.COLORS[color])
        ax.text(962, y_top - 118, d, ha="left", va="top", fontsize=21.5,
                color=nm.COLORS["muted"], linespacing=1.5)

    nm.header(fig, text["title"], text["sub"])
    nm.footer(fig, text["note"])
    return fig

build_figure(state_flows(LANG), LANG)
```
