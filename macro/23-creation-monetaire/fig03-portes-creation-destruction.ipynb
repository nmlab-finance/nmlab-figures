```python
LANG = "fr"   # "fr" ou "en" — langue des libellés / label language

# Récupère puis active le style partagé NMLab (thème sombre + police Inter).
# Fetch and activate the shared NMLab style (dark theme + Inter font).
import urllib.request

urllib.request.urlretrieve("https://raw.githubusercontent.com/nmlab-finance/nmlab-figures/main/nmlab_style.py", "nmlab_style.py")
import nmlab_style as nm

nm.setup()


def doors(lang: str) -> tuple[list[tuple[str, str]], list[tuple[str, str]]]:
    """Portes de création et de destruction de la monnaie : (titre, détail).
    Doors of money creation and destruction: (title, detail)."""
    if lang == "fr":
        create = [
            ("Une banque prête", "le crédit crée le dépôt (chapitre 22)"),
            ("Une banque achète un titre", "le vendeur non bancaire est payé en dépôt neuf"),
            ("La banque centrale achète (QE)", "via un fonds : dépôt et réserves créés ensemble"),
            ("L'État dépense plus qu'il ne taxe", "si banques ou banque centrale absorbent la dette"),
            ("Des devises entrent", "l'exportateur convertit ses dollars en euros"),
        ]
        destroy = [
            ("Un crédit est remboursé", "le principal éteint le dépôt (chapitre 22)"),
            ("Une banque vend un titre", "l'acheteur non bancaire paie de son dépôt, détruit"),
            ("La banque centrale réduit (QT)", "titres non remplacés : réserves et dépôts refluent"),
            ("L'impôt dépasse la dépense", "l'excédent draine les dépôts vers le Trésor"),
            ("L'épargne se fige en titres longs", "les dépôts basculent hors de la monnaie (M3)"),
        ]
    else:
        create = [
            ("A bank lends", "the loan creates the deposit (chapter 22)"),
            ("A bank buys a security", "the non-bank seller is paid with a new deposit"),
            ("The central bank buys (QE)", "via a fund: deposit and reserves created together"),
            ("The state spends more than it taxes", "if banks or the central bank absorb the debt"),
            ("Foreign currency flows in", "the exporter converts dollars into euro deposits"),
        ]
        destroy = [
            ("A loan is repaid", "the principal extinguishes the deposit (chapter 22)"),
            ("A bank sells a security", "the non-bank buyer pays out of a deposit, destroyed"),
            ("The central bank shrinks (QT)", "bonds mature unreplaced: reserves and deposits ebb"),
            ("Taxes exceed spending", "the surplus drains deposits into the Treasury"),
            ("Savings lock into long-term paper", "deposits shift out of money (M3)"),
        ]
    return create, destroy


from matplotlib.figure import Figure

LABELS = {
    "fr": dict(
        title="Les portes de la création — et de la destruction",
        sub="La masse monétaire est un solde : chaque écriture qui crée un dépôt a sa jumelle qui l'éteint.",
        left="La monnaie apparaît quand…", right="…et disparaît quand",
        note="Le chapitre 22 n'ouvrait que la première porte. La masse monétaire additionne toutes les autres —\n"
             "banques, banque centrale, État, extérieur — créations moins destructions."),
    "en": dict(
        title="The doors of creation — and destruction",
        sub="The money stock is a balance: every entry that creates a deposit has a twin that extinguishes it.",
        left="Money appears when…", right="…and disappears when",
        note="Chapter 22 opened only the first door. The money stock adds up all the others — banks, the central\n"
             "bank, the state, the rest of the world — creations minus destructions."),
}

def build_figure(doors_: tuple[list[tuple[str, str]], list[tuple[str, str]]], lang: str) -> Figure:
    """Deux colonnes de cinq cartes : création (bleu) à gauche, destruction (rose) à droite."""
    text = LABELS[lang]
    fig = nm.figure(height_px=1230)
    ax = nm.blank_axes(fig)

    create, destroy = doors_
    col_w, x_left, x_right = 795, 62, 890
    top, card_h, gap = 1000, 148, 22

    for x, head, rows, color in ((x_left, text["left"], create, "blue"),
                                 (x_right, text["right"], destroy, "rose")):
        ax.text(x + 6, top + 44, head, ha="left", va="center", fontsize=26,
                fontweight="bold", color=nm.COLORS[color])
        for i, (t, d) in enumerate(rows):
            y = top - 40 - (i + 1) * card_h - i * gap + card_h
            nm.card(ax, x, y - card_h, col_w, card_h, edge=nm.COLORS[color], lw=2.2, radius=20)
            ax.text(x + 38, y - 50, t, ha="left", va="center", fontsize=24.5,
                    fontweight="bold", color=nm.COLORS["text"])
            ax.text(x + 38, y - 104, d, ha="left", va="center", fontsize=19.5,
                    color=nm.COLORS["muted"])

    nm.header(fig, text["title"], text["sub"])
    nm.footer(fig, text["note"])
    return fig

build_figure(doors(LANG), LANG)
```
