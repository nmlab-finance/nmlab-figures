```python
LANG = "fr"   # "fr" ou "en" — langue des libellés / label language

# Récupère puis active le style partagé NMLab (thème sombre + police Inter).
# Fetch and activate the shared NMLab style (dark theme + Inter font).
import urllib.request

urllib.request.urlretrieve("https://raw.githubusercontent.com/nmlab-finance/nmlab-figures/main/nmlab_style.py", "nmlab_style.py")
import nmlab_style as nm

nm.setup()


def cascade(lang: str) -> list[tuple[str, float]]:
    """Les cinq premiers dépôts de la cascade à 10 % de réserves obligatoires.
    The first five deposits of the cascade with 10% required reserves."""
    label = "Dépôt" if lang == "fr" else "Deposit"
    return [(f"{label} {i + 1}", 100 * 0.9 ** i) for i in range(5)]


from matplotlib.figure import Figure

LABELS = {
    "fr": dict(
        title="Le multiplicateur du manuel : 100 deviennent 1 000",
        sub="Chaque dépôt est re-prêté à 90 %, la réserve de 10 % reste au coffre — la série somme à 1/r.",
        lent="re-prêté (90 %)", reserve="réserve (10 %)", dots="… et ainsi de suite",
        total="Somme des dépôts\n100 × 1 ⁄ 10 % = 1 000",
        note="La mécanique enseignée depuis Phillips (1920) et les manuels d'après-guerre : la banque centrale fixe\n"
             "les réserves, le système « multiplie ». Élégant — reste à savoir si les banques travaillent ainsi."),
    "en": dict(
        title="The textbook multiplier: 100 becomes 1,000",
        sub="Each deposit is re-lent at 90%, the 10% reserve stays in the vault — the series sums to 1/r.",
        lent="re-lent (90%)", reserve="reserve (10%)", dots="… and so on",
        total="Sum of deposits\n100 × 1 ⁄ 10% = 1,000",
        note="The mechanics taught since Phillips (1920) and the post-war textbooks: the central bank sets reserves,\n"
             "the system \"multiplies\" them. Elegant — the question is whether banks actually work that way."),
}

def build_figure(rows: list[tuple[str, float]], lang: str) -> Figure:
    """Cascade de barres 100, 90, 81… scindées re-prêté/réserve, et carte de la somme."""
    text = LABELS[lang]
    fig = nm.figure(height_px=1120)
    ax = nm.blank_axes(fig)

    x0, full_w = 300, 1000                 # 100 unités = 1000 px
    top, row_h, gap = 880, 96, 34
    for i, (label, value) in enumerate(rows):
        y = top - i * (row_h + gap)
        w = full_w * value / 100
        ax.add_patch(plt.Rectangle((x0, y - row_h), w * 0.9, row_h,
                                   facecolor=nm.COLORS["blue"], edgecolor="none", zorder=2))
        ax.add_patch(plt.Rectangle((x0 + w * 0.9, y - row_h), w * 0.1, row_h,
                                   facecolor=nm.COLORS["amber"], edgecolor="none", zorder=2))
        ax.text(x0 - 24, y - row_h / 2, label, ha="right", va="center", fontsize=22,
                color=nm.COLORS["muted"])
        ax.text(x0 + w + 20, y - row_h / 2, f"{value:.1f}".replace(".0", "").replace(".", "," if lang == "fr" else "."),
                ha="left", va="center", fontsize=22.5, fontweight="bold", color=nm.COLORS["text"])
    y_dots = top - 5 * (row_h + gap)
    ax.text(x0, y_dots - 30, text["dots"], ha="left", va="center", fontsize=22,
            fontstyle="italic", color=nm.COLORS["muted"])

    # Légende re-prêté / réserve, au-dessus des barres.
    for dx, color, lab in ((0, "blue", text["lent"]), (330, "amber", text["reserve"])):
        ax.add_patch(plt.Rectangle((x0 + dx, top + 44), 34, 26,
                                   facecolor=nm.COLORS[color], edgecolor="none"))
        ax.text(x0 + dx + 48, top + 57, lab, ha="left", va="center", fontsize=20.5,
                color=nm.COLORS["muted"])

    # La somme.
    nm.card(ax, 1330, y_dots - 130, 360, 190, edge=nm.COLORS["blue"], lw=2.6, radius=22)
    ax.text(1510, y_dots - 35, text["total"], ha="center", va="center", fontsize=23.5,
            fontweight="bold", color=nm.COLORS["blue"], linespacing=1.6)

    nm.header(fig, text["title"], text["sub"])
    nm.footer(fig, text["note"])
    return fig

import matplotlib.pyplot as plt

build_figure(cascade(LANG), LANG)
```
