```python
LANG = "fr"   # "fr" ou "en" — langue des libellés / label language

# Récupère puis active le style partagé NMLab (thème sombre + police Inter).
# Fetch and activate the shared NMLab style (dark theme + Inter font).
import urllib.request

urllib.request.urlretrieve("https://raw.githubusercontent.com/nmlab-finance/nmlab-figures/main/nmlab_style.py", "nmlab_style.py")
import nmlab_style as nm

nm.setup()


def stories(lang: str) -> list[dict]:
    """Les deux récits de la création de monnaie : manuel et réalité, localisés.
    The two stories of money creation: textbook and reality, localized."""
    if lang == "fr":
        return [
            dict(head="Le manuel — la flèche descend", color="rose",
                 steps=["Réserves", "Prêts", "Dépôts"],
                 sub="La banque centrale fixe la quantité de réserves ; les banques la « multiplient » en dépôts."),
            dict(head="La réalité — la flèche remonte", color="blue",
                 steps=["Prêts", "Dépôts", "Réserves"],
                 sub="La banque prête d'abord ; les réserves suivent, fournies au prix fixé par la banque centrale."),
        ]
    return [
        dict(head="The textbook — the arrow runs down", color="rose",
             steps=["Reserves", "Loans", "Deposits"],
             sub="The central bank sets the quantity of reserves; banks \"multiply\" it into deposits."),
        dict(head="Reality — the arrow runs back up", color="blue",
             steps=["Loans", "Deposits", "Reserves"],
             sub="The bank lends first; reserves follow, supplied at the price the central bank sets."),
    ]

def holmes_quote(lang: str) -> str:
    """La phrase d'Alan Holmes (Fed de New York, 1969). Alan Holmes's 1969 sentence."""
    if lang == "fr":
        return ("« Dans le monde réel, les banques accordent leurs crédits, créant les dépôts au passage,\n"
                "et cherchent les réserves ensuite. » — Alan Holmes, Fed de New York, 1969")
    return ("\"In the real world, banks extend credit, creating deposits in the process,\n"
            "and look for the reserves later.\" — Alan Holmes, New York Fed, 1969")


from matplotlib.figure import Figure

LABELS = {
    "fr": dict(
        title="Deux récits, une flèche inversée",
        sub="Même comptabilité, causalité opposée : la question du chapitre tient dans le sens de la flèche.",
        note="Celui qui pilotait au quotidien les opérations de marché de la Fed décrivait déjà, en 1969, la flèche\n"
             "qui remonte. Source : A. Holmes, « Operational Constraints on the Stabilization of Money Supply Growth »."),
    "en": dict(
        title="Two stories, one reversed arrow",
        sub="Same accounting, opposite causality: the chapter's question fits in the direction of the arrow.",
        note="The man running the Fed's day-to-day market operations had already described the upward arrow in 1969.\n"
             "Source: A. Holmes, \"Operational Constraints on the Stabilization of Money Supply Growth.\""),
}

def build_figure(stories_: list[dict], quote: str, lang: str) -> Figure:
    """Deux bandes récit (chips reliées par flèches) puis la carte-citation de Holmes."""
    text = LABELS[lang]
    fig = nm.figure(height_px=1120)
    ax = nm.blank_axes(fig)

    bands = [(620, 900), (300, 580)]
    for spec, (b, t) in zip(stories_, bands):
        nm.card(ax, 62, b, 1623, t - b, edge=nm.COLORS[spec["color"]], lw=2.6, radius=24)
        ax.text(104, t - 52, spec["head"], ha="left", va="center", fontsize=27,
                fontweight="bold", color=nm.COLORS[spec["color"]])
        chip_w, chip_h, x_start, y_chip = 250, 78, 104, b + 92
        for i, step in enumerate(spec["steps"]):
            x = x_start + i * (chip_w + 120)
            nm.card(ax, x, y_chip, chip_w, chip_h, edge=nm.COLORS["edge"],
                    fill=nm.COLORS["bg"], lw=2, radius=18)
            ax.text(x + chip_w / 2, y_chip + chip_h / 2, step, ha="center", va="center",
                    fontsize=23, fontweight="bold", color=nm.COLORS["text"])
            if i < 2:
                ax.annotate("", xy=(x + chip_w + 108, y_chip + chip_h / 2),
                            xytext=(x + chip_w + 12, y_chip + chip_h / 2),
                            arrowprops=dict(arrowstyle="-|>", lw=3,
                                            color=nm.COLORS[spec["color"]], mutation_scale=24))
        ax.text(104, b + 44, spec["sub"], ha="left", va="center",
                fontsize=20, color=nm.COLORS["muted"])

    nm.card(ax, 62, 100, 1623, 150, edge=nm.COLORS["amber"], lw=2.4, radius=22)
    ax.text(873, 175, quote, ha="center", va="center", fontsize=22.5,
            fontstyle="italic", color=nm.COLORS["amber"], linespacing=1.6)

    nm.header(fig, text["title"], text["sub"])
    nm.footer(fig, text["note"])
    return fig

build_figure(stories(LANG), holmes_quote(LANG), LANG)
```
