```python
LANG = "fr"   # "fr" ou "en" — langue des libellés / label language

# Récupère puis active le style partagé NMLab (thème sombre + police Inter).
# Fetch and activate the shared NMLab style (dark theme + Inter font).
import urllib.request

urllib.request.urlretrieve("https://raw.githubusercontent.com/nmlab-finance/nmlab-figures/main/nmlab_style.py", "nmlab_style.py")
import nmlab_style as nm

nm.setup()


def aftermath(lang: str) -> list[tuple[str, str]]:
    """Trois faits d'après-lettre : (titre, détail), localisés.
    Three post-letter facts: (title, detail), localized."""
    if lang == "fr":
        return [
            ("Base monétaire : ×2", "encore doublée entre la lettre\net 2014 — après QE2 puis QE3"),
            ("Inflation : +1,7 % par an", "IPC américain, moyenne\n2010-2015 — sous la cible"),
            ("L'or retombe", "porté à 1 900 dollars l'once\nen 2011, sous 1 100 fin 2015"),
        ]
    return [
        ("Monetary base: ×2", "doubled again between the\nletter and 2014 — QE2, then QE3"),
        ("Inflation: +1.7% a year", "US CPI, 2010-2015 average —\nbelow target"),
        ("Gold falls back", "carried to 1,900 dollars an ounce\nin 2011, below 1,100 by end-2015"),
    ]


from matplotlib.figure import Figure

LABELS = {
    "fr": dict(
        title="Novembre 2010 : la lettre — et la décennie d'après",
        sub="Vingt-trois investisseurs, économistes et essayistes demandent à la Fed d'arrêter ses achats d'actifs.",
        letter_head="« Open Letter to Ben Bernanke »",
        letter_meta="Wall Street Journal, 15 novembre 2010",
        quote="« Le programme d'achats d'actifs devrait\nêtre reconsidéré : il risque une\ndépréciation de la monnaie\net de l'inflation. »",
        signers="Parmi les signataires : Cliff Asness,\nNiall Ferguson, James Grant,\nPaul Singer, John Taylor…",
        after_head="Ce qui a suivi",
        note="Le raisonnement de la lettre reposait sur le multiplicateur : des réserves multipliées finissent en\n"
             "inflation. Ce chapitre instruit ce procès. Sources : WSJ (15/11/2010) ; FRED ; cours de l'or LBMA."),
    "en": dict(
        title="November 2010: the letter — and the decade after",
        sub="Twenty-three investors, economists and essayists ask the Fed to stop its asset purchases.",
        letter_head="\"Open Letter to Ben Bernanke\"",
        letter_meta="Wall Street Journal, November 15, 2010",
        quote="\"The planned asset purchases should be\nreconsidered: they risk currency\ndebasement and inflation.\"",
        signers="Among the signatories: Cliff Asness,\nNiall Ferguson, James Grant,\nPaul Singer, John Taylor…",
        after_head="What followed",
        note="The letter's reasoning rested on the multiplier: multiplied reserves must end in inflation. This\n"
             "chapter puts that on trial. Sources: WSJ (Nov. 15, 2010); FRED; LBMA gold prices."),
}

def build_figure(facts: list[tuple[str, str]], lang: str) -> Figure:
    """Carte-lettre à gauche (citation), trois cartes de faits à droite."""
    text = LABELS[lang]
    fig = nm.figure(height_px=1120)
    ax = nm.blank_axes(fig)

    # La lettre.
    nm.card(ax, 62, 200, 760, 690, edge=nm.COLORS["rose"], lw=2.6, radius=24)
    ax.text(102, 826, text["letter_head"], ha="left", va="center", fontsize=27,
            fontweight="bold", color=nm.COLORS["text"])
    ax.text(102, 770, text["letter_meta"], ha="left", va="center", fontsize=20.5,
            color=nm.COLORS["muted"])
    ax.plot([102, 782], [734, 734], color=nm.COLORS["edge"], lw=2)
    ax.text(102, 664, text["quote"], ha="left", va="top", fontsize=24.5,
            fontstyle="italic", color=nm.COLORS["rose"], linespacing=1.65)
    ax.text(102, 386, text["signers"], ha="left", va="top", fontsize=20,
            color=nm.COLORS["muted"], linespacing=1.5)

    # Les trois faits.
    ax.text(930, 852, text["after_head"], ha="left", va="center", fontsize=26,
            fontweight="bold", color=nm.COLORS["text"])
    for i, (t, d) in enumerate(facts):
        y_top = 790 - i * 230
        nm.card(ax, 924, y_top - 208, 760, 208, edge=nm.COLORS["edge"], lw=2.2, radius=22)
        ax.text(962, y_top - 54, t, ha="left", va="center", fontsize=25,
                fontweight="bold", color=nm.COLORS["blue"])
        ax.text(962, y_top - 106, d, ha="left", va="top", fontsize=20.5,
                color=nm.COLORS["muted"], linespacing=1.4)

    nm.header(fig, text["title"], text["sub"])
    nm.footer(fig, text["note"])
    return fig

build_figure(aftermath(LANG), LANG)
```
