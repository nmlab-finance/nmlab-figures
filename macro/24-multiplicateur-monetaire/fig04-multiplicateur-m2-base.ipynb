```python
LANG = "fr"   # "fr" ou "en" — langue des libellés / label language

# Récupère puis active le style partagé NMLab (thème sombre + police Inter).
# Fetch and activate the shared NMLab style (dark theme + Inter font).
import urllib.request

urllib.request.urlretrieve("https://raw.githubusercontent.com/nmlab-finance/nmlab-figures/main/nmlab_style.py", "nmlab_style.py")
import nmlab_style as nm

nm.setup()


from pandas import Series


def load_multiplier() -> Series:
    """Ratio M2 / base monétaire, mensuel depuis 1959 (FRED : M2SL, BOGMBASE).
    The M2-to-monetary-base ratio, monthly since 1959 (FRED: M2SL, BOGMBASE)."""
    m2 = nm.load_fred("M2SL")
    base = nm.load_fred("BOGMBASE")
    return (m2 / base).dropna()


from matplotlib.figure import Figure
from pandas import Series

LABELS = {
    "fr": dict(
        title="Le « multiplicateur » américain, 1959-2026",
        sub="Ratio M2 / base monétaire — si c'était une constante de la nature, cette courbe serait plate.",
        hi="11,8 — févr. 1985", pre="8,9 — fin 2007", lo="2,8 — août 2014",
        qe="nov. 2008 :\npremier QE",
        note="Un « multiplicateur » divisé par trois en six ans, sans pénurie de monnaie ni hyperinflation : un\n"
             "coefficient de comportement, pas une loi. Source : FRED (M2SL, BOGMBASE)."),
    "en": dict(
        title="The U.S. \"multiplier\", 1959-2026",
        sub="M2 divided by the monetary base — if this were a constant of nature, the line would be flat.",
        hi="11.8 — Feb. 1985", pre="8.9 — late 2007", lo="2.8 — Aug. 2014",
        qe="Nov. 2008:\nfirst QE",
        note="A \"multiplier\" divided by three in six years, with no money shortage and no hyperinflation: a\n"
             "behavioural ratio, not a law. Source: FRED (M2SL, BOGMBASE)."),
}

def build_figure(mult: Series, lang: str) -> Figure:
    """Courbe du ratio avec trois annotations (sommet 1985, fin 2007, creux 2014) et jalon QE."""
    import pandas as pd

    text = LABELS[lang]
    fig = nm.figure(height_px=1045)
    ax = nm.axes(fig, left=0.05)
    ax.plot(mult.index, mult, color=nm.COLORS["blue"], lw=3, zorder=3)
    ax.set_ylim(0, 14.4)
    ax.set_xlim(pd.Timestamp("1957-06-01"), pd.Timestamp("2028-06-01"))

    qe = pd.Timestamp("2008-11-01")
    ax.axvline(qe, color=nm.COLORS["edge"], lw=2, linestyle=(0, (5, 5)), zorder=1)
    ax.text(qe - pd.Timedelta(days=300), 0.65, text["qe"], ha="right", va="bottom",
            fontsize=19, color=nm.COLORS["muted"], linespacing=1.4)

    hi = mult.idxmax()
    ax.scatter([hi], [mult.max()], s=90, color=nm.COLORS["amber"], zorder=4)
    ax.text(hi, mult.max() + 0.55, text["hi"], ha="center", va="bottom", fontsize=21,
            fontweight="bold", color=nm.COLORS["amber"])

    pre = pd.Timestamp("2007-12-01")
    ax.scatter([pre], [mult.loc[pre]], s=90, color=nm.COLORS["teal"], zorder=4)
    ax.text(pre - pd.Timedelta(days=420), mult.loc[pre] + 0.5, text["pre"], ha="right",
            va="bottom", fontsize=21, fontweight="bold", color=nm.COLORS["teal"])

    lo = mult.loc["2010":].idxmin()
    ax.scatter([lo], [mult.loc[lo]], s=90, color=nm.COLORS["rose"], zorder=4)
    ax.text(lo, mult.loc[lo] - 0.6, text["lo"], ha="center", va="top", fontsize=21,
            fontweight="bold", color=nm.COLORS["rose"])

    nm.header(fig, text["title"], text["sub"])
    nm.footer(fig, text["note"])
    return fig

build_figure(load_multiplier(), LANG)
```
