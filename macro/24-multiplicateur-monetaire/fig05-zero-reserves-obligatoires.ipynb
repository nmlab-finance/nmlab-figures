```python
LANG = "fr"   # "fr" ou "en" — langue des libellés / label language

# Récupère puis active le style partagé NMLab (thème sombre + police Inter).
# Fetch and activate the shared NMLab style (dark theme + Inter font).
import urllib.request

urllib.request.urlretrieve("https://raw.githubusercontent.com/nmlab-finance/nmlab-figures/main/nmlab_style.py", "nmlab_style.py")
import nmlab_style as nm

nm.setup()


def zero_reserve(lang: str) -> list[tuple[str, str]]:
    """Six systèmes bancaires sans réserves obligatoires : (pays, précision).
    Six banking systems with no required reserves: (country, detail)."""
    if lang == "fr":
        return [
            ("Canada", "0 % depuis 1994"),
            ("Royaume-Uni", "0 % — régime volontaire"),
            ("Australie", "0 %"),
            ("Nouvelle-Zélande", "0 % depuis 1985"),
            ("Suède", "0 % depuis 1994"),
            ("États-Unis", "0 % depuis mars 2020"),
        ]
    return [
        ("Canada", "0% since 1994"),
        ("United Kingdom", "0% — voluntary regime"),
        ("Australia", "0%"),
        ("New Zealand", "0% since 1985"),
        ("Sweden", "0% since 1994"),
        ("United States", "0% since March 2020"),
    ]


from matplotlib.figure import Figure

LABELS = {
    "fr": dict(
        title="r = 0 : le multiplicateur infini n'a pas eu lieu",
        sub="Avec zéro réserve obligatoire, la formule 1 ⁄ r prédit une création sans borne. Elle n'existe nulle part.",
        euro="Zone euro : 1 % — et le crédit n'y est pas ×100. Le frein est ailleurs :\n"
             "capital, rentabilité, demande, liquidité (chapitre 22).",
        note="Les réserves obligatoires subsistent surtout comme outil technique de pilotage des taux courts —\n"
             "pas comme digue quantitative. Sources : banques centrales nationales ; Fed (15 mars 2020)."),
    "en": dict(
        title="r = 0: the infinite multiplier never happened",
        sub="With zero required reserves, the 1 ⁄ r formula predicts boundless creation. It exists nowhere.",
        euro="Euro area: 1% — and credit there is not ×100. The brake lies elsewhere:\n"
             "capital, profitability, demand, liquidity (chapter 22).",
        note="Where reserve requirements survive, they are mostly a technical tool for steering short rates —\n"
             "not a quantitative dike. Sources: national central banks; the Fed (March 15, 2020)."),
}

def build_figure(countries: list[tuple[str, str]], lang: str) -> Figure:
    """Grille 3 × 2 de cartes pays, puis carte zone euro pleine largeur."""
    text = LABELS[lang]
    fig = nm.figure(height_px=1120)
    ax = nm.blank_axes(fig)

    card_w, card_h, gap_x, gap_y, x0, top = 515, 220, 39, 36, 62, 880
    for i, (country, detail) in enumerate(countries):
        x = x0 + (i % 3) * (card_w + gap_x)
        y = top - (i // 3) * (card_h + gap_y)
        nm.card(ax, x, y - card_h, card_w, card_h, edge=nm.COLORS["blue"], lw=2.4, radius=22)
        ax.text(x + 40, y - 66, country, ha="left", va="center", fontsize=27,
                fontweight="bold", color=nm.COLORS["text"])
        ax.text(x + 40, y - 140, detail, ha="left", va="center", fontsize=22,
                color=nm.COLORS["muted"])

    nm.card(ax, 62, 170, 1623, 180, edge=nm.COLORS["amber"], lw=2.4, radius=22)
    ax.text(873, 260, text["euro"], ha="center", va="center", fontsize=23,
            fontweight="bold", color=nm.COLORS["amber"], linespacing=1.55)

    nm.header(fig, text["title"], text["sub"])
    nm.footer(fig, text["note"])
    return fig

build_figure(zero_reserve(LANG), LANG)
```
