```python
LANG = "fr"   # "fr" ou "en" — langue des libellés / label language

# Récupère puis active le style partagé NMLab (thème sombre + police Inter).
# Fetch and activate the shared NMLab style (dark theme + Inter font).
import urllib.request

urllib.request.urlretrieve("https://raw.githubusercontent.com/nmlab-finance/nmlab-figures/main/nmlab_style.py", "nmlab_style.py")
import nmlab_style as nm

nm.setup()


from pandas import DataFrame


def load_m2_cpi() -> DataFrame:
    """M2 et IPC américains en glissement annuel (%), mensuels depuis 2015
    (FRED : M2SL ; CPIAUCNS, série non désaisonnalisée du fameux « 9,1 % »).
    US M2 and CPI, % year over year, monthly since 2015 (FRED: M2SL; CPIAUCNS)."""
    import pandas as pd

    m2 = nm.load_fred("M2SL").pct_change(12) * 100
    cpi = nm.load_fred("CPIAUCNS").pct_change(12) * 100
    frame = pd.concat({"m2": m2, "cpi": cpi}, axis=1).dropna()
    return frame.loc["2015":]


from matplotlib.figure import Figure
from pandas import DataFrame

LABELS = {
    "fr": dict(
        title="La monnaie large parle — quand on l'écoute",
        sub="M2 et prix à la consommation, glissements annuels : le pic de 2021 précède celui de 2022 de seize mois.",
        m2="M2", cpi="IPC",
        peak_m2="+26,8 % — févr. 2021",
        peak_cpi="+9,1 % — juin 2022",
        trough_m2="−4,6 % — avril 2023",
        note="La leçon en creux du multiplicateur : la base était muette, la monnaie LARGE ne l'était pas — surtout\n"
             "une fois logée chez les ménages. Source : FRED (M2SL ; CPIAUCNS, non désaisonnalisé)."),
    "en": dict(
        title="Broad money speaks — when you listen",
        sub="M2 and consumer prices, year over year: the 2021 peak precedes the 2022 one by sixteen months.",
        m2="M2", cpi="CPI",
        peak_m2="+26.8% — Feb. 2021",
        peak_cpi="+9.1% — June 2022",
        trough_m2="−4.6% — Apr. 2023",
        note="The multiplier's lesson in reverse: the base was mute, BROAD money was not — above all once it sat in\n"
             "household accounts. Source: FRED (M2SL; CPIAUCNS, not seasonally adjusted)."),
}

def build_figure(frame: DataFrame, lang: str) -> Figure:
    """Deux courbes (M2, IPC) sur un même axe %, pics et creux annotés."""
    import pandas as pd

    text = LABELS[lang]
    fig = nm.figure(height_px=1045)
    ax = nm.axes(fig, left=0.056)
    ax.plot(frame.index, frame["m2"], color=nm.COLORS["blue"], lw=3.2, zorder=3, label=text["m2"])
    ax.plot(frame.index, frame["cpi"], color=nm.COLORS["rose"], lw=3.2, zorder=3, label=text["cpi"])
    ax.axhline(0, color=nm.COLORS["muted"], lw=1.6, alpha=0.85, zorder=2)
    ax.set_ylim(-8, 31)
    ax.set_xlim(pd.Timestamp("2014-08-01"), pd.Timestamp("2027-02-01"))

    peak_m2 = frame["m2"].idxmax()
    ax.scatter([peak_m2], [frame["m2"].max()], s=90, color=nm.COLORS["blue"], zorder=4)
    ax.text(peak_m2, frame["m2"].max() + 1.1, text["peak_m2"], ha="center", va="bottom",
            fontsize=21, fontweight="bold", color=nm.COLORS["blue"])

    peak_cpi = frame["cpi"].idxmax()
    ax.scatter([peak_cpi], [frame["cpi"].max()], s=90, color=nm.COLORS["rose"], zorder=4)
    ax.text(peak_cpi + pd.Timedelta(days=120), frame["cpi"].max() + 1.1, text["peak_cpi"],
            ha="left", va="bottom", fontsize=21, fontweight="bold", color=nm.COLORS["rose"])

    trough_m2 = frame["m2"].idxmin()
    ax.scatter([trough_m2], [frame["m2"].min()], s=90, color=nm.COLORS["blue"], zorder=4)
    ax.text(trough_m2, frame["m2"].min() - 1.2, text["trough_m2"], ha="center", va="top",
            fontsize=21, fontweight="bold", color=nm.COLORS["blue"])

    legend = ax.legend(loc="upper right", fontsize=20, frameon=True)
    legend.get_frame().set_facecolor(nm.COLORS["bg"])
    legend.get_frame().set_edgecolor(nm.COLORS["edge"])

    nm.header(fig, text["title"], text["sub"])
    nm.footer(fig, text["note"])
    return fig

build_figure(load_m2_cpi(), LANG)
```
